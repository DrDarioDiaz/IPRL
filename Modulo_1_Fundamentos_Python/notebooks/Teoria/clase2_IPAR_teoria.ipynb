```python
# Estructuras de Datos Avanzadas en Python para Aprendizaje por Refuerzo
# Dr. Darío Ezequiel Díaz
# Marzo 2025

import numpy as np
import matplotlib.pyplot as plt
import random

# ==================================================================================
# 1. LISTAS EN PYTHON
# ==================================================================================
print("=" * 80)
print("LISTAS EN PYTHON")
print("=" * 80)

# Creación de diferentes tipos de listas
numeros = [1, 2, 3, 4, 5]
datos_varios = [42, "hola", 3.1416, False, [10, 20]]
lista_vacia = []

print("Lista de números:", numeros)
print("Lista heterogénea:", datos_varios)
print("Lista vacía:", lista_vacia)

# Acceso a elementos
colores = ["rojo", "verde", "azul", "amarillo"]

# Índice positivo: comienza desde 0
primer_color = colores[0]
print("\nPrimer color (índice 0):", primer_color)

# Índice negativo: comienza desde -1 (último elemento)
ultimo_color = colores[-1]
print("Último color (índice -1):", ultimo_color)

# Slicing (rebanadas)
print("Primeros dos colores:", colores[0:2])
print("Últimos dos colores:", colores[-2:])
print("Cada segundo color:", colores[::2])

# Modificación de listas
numeros = [10, 20, 30, 40, 50]
print("\nLista original:", numeros)

# Modificar elemento existente
numeros[2] = 35
print("Después de modificar índice 2:", numeros)

# Añadir elementos
numeros.append(60)
print("Después de append(60):", numeros)

# Insertar en posición específica
numeros.insert(1, 15)
print("Después de insert(1, 15):", numeros)

# Eliminar elementos por valor
numeros.remove(15)
print("Después de remove(15):", numeros)

# Eliminar por índice
del numeros[0]
print("Después de del numeros[0]:", numeros)

# Extraer elemento con retorno
elemento = numeros.pop(2)
print(f"Elemento extraído: {elemento}, Lista resultante: {numeros}")

# Métodos de ordenamiento y búsqueda
numeros = [42, 13, 7, 24, 36, 5]
print("\nLista original:", numeros)

# Ordenamiento in-place (modifica la lista original)
numeros_copia = numeros.copy()
numeros_copia.sort()
print("Lista ordenada (in-place):", numeros_copia)

# Ordenamiento inverso
numeros_copia = numeros.copy()
numeros_copia.sort(reverse=True)
print("Lista ordenada descendente:", numeros_copia)

# Ordenamiento sin modificar la original (devuelve una nueva lista)
numeros_ordenados = sorted(numeros)
print("Lista original (sin cambios):", numeros)
print("Nueva lista ordenada (sorted):", numeros_ordenados)

# Inversión
numeros_copia = numeros.copy()
numeros_copia.reverse()
print("Lista invertida:", numeros_copia)

# Ordenamiento de listas heterogéneas
lista_mixta = [42, "cadena", True, 3.14159]
print("\nLista mixta original:", lista_mixta)

# Ordenamiento con clave de comparación
lista_ordenada = sorted(lista_mixta, key=str)
print("Lista ordenada por representación string:", lista_ordenada)

# Intento de ordenar sin clave
try:
   lista_mixta_copia = lista_mixta.copy()
   lista_mixta_copia.sort()
except TypeError as e:
   print("Error al ordenar sin clave:", e)

# ==================================================================================
# 2. TUPLAS EN PYTHON
# ==================================================================================
print("\n" + "=" * 80)
print("TUPLAS EN PYTHON")
print("=" * 80)

# Creación de tuplas
coordenadas = (10, 20)
persona = ("Ana", 28, "Ingeniera")
singleton = (42,)  # Nota la coma, necesaria para tuplas de un elemento
tupla_implicita = 1, 2, 3  # Equivalente a (1, 2, 3)

print("Coordenadas:", coordenadas)
print("Datos de persona:", persona)
print("Tupla con un solo elemento:", singleton)
print("Tupla creada implícitamente:", tupla_implicita)

# Inmutabilidad
tupla = (10, 20, 30)
print("\nTupla original:", tupla)

# Intentar modificar un elemento genera error
try:
   tupla[0] = 99
except TypeError as e:
   print("Error al intentar modificar tupla:", e)

# Podemos crear nuevas tuplas basadas en existentes
tupla_original = (1, 2, 3)
nueva_tupla = tupla_original + (4, 5)
subtupla = tupla_original[0:2]
tupla_repetida = tupla_original * 2

print("\nTupla original:", tupla_original)
print("Nueva tupla (concatenación):", nueva_tupla)
print("Subtupla (slicing):", subtupla)
print("Tupla repetida:", tupla_repetida)

# Desempaquetado de tuplas
# Desempaquetado básico
tupla = (100, 200, 300)
x, y, z = tupla
print(f"\nx = {x}, y = {y}, z = {z}")

# Aplicación práctica: intercambio de variables
a, b = 10, 20
print(f"Antes del intercambio: a = {a}, b = {b}")
a, b = b, a  # Intercambio mediante desempaquetado
print(f"Después del intercambio: a = {a}, b = {b}")

# Desempaquetado con *rest
numeros = (1, 2, 3, 4, 5)
primero, *resto = numeros
print(f"Primer elemento: {primero}, Resto: {resto}")

# Ignorando valores con _
datos = ("Ana", 28, "Ingeniera", "Madrid")
nombre, edad, _, ciudad = datos
print(f"Nombre: {nombre}, Edad: {edad}, Ciudad: {ciudad}")

# ==================================================================================
# 3. DICCIONARIOS EN PYTHON
# ==================================================================================
print("\n" + "=" * 80)
print("DICCIONARIOS EN PYTHON")
print("=" * 80)

# Creación de diccionarios
estudiante = {
   "nombre": "Ana García",
   "edad": 22,
   "carrera": "Ingeniería Informática",
   "promedio": 8.7
}

# Diccionario vacío
diccionario_vacio = {}

# Creación mediante la función dict()
coordenadas = dict(x=10, y=20, z=30)

# Desde secuencia de pares clave-valor
items = [("a", 1), ("b", 2), ("c", 3)]
letras_numeros = dict(items)

print("Estudiante:", estudiante)
print("Coordenadas:", coordenadas)
print("Letras y números:", letras_numeros)

# Operaciones fundamentales con diccionarios
diccionario = {"a": 1, "b": 2, "c": 3}
print("\nDiccionario original:", diccionario)

# Acceso por clave
valor_b = diccionario["b"]
print("Valor de 'b':", valor_b)

# Acceso seguro con get (para claves que quizás no existan)
valor_d = diccionario.get("d", 0)  # Devuelve 0 si "d" no existe
print("Valor de 'd' (por defecto 0):", valor_d)

# Insertar o modificar
diccionario["d"] = 4  # Añade nuevo par
print("Después de añadir 'd':", diccionario)

diccionario["a"] = 10  # Modifica valor existente
print("Después de modificar 'a':", diccionario)

# Eliminar pares
del diccionario["b"]
print("Después de eliminar 'b':", diccionario)

# Eliminar y obtener valor con pop
valor_c = diccionario.pop("c")
print(f"Valor extraído 'c': {valor_c}, Diccionario resultante: {diccionario}")

# Vaciar diccionario
diccionario_copia = diccionario.copy()
diccionario_copia.clear()
print("Diccionario después de clear():", diccionario_copia)

# Verificación y recuperación de datos
diccionario = {"a": 1, "b": 2, "c": 3}

# Verificar existencia de clave
existe_b = "b" in diccionario
existe_x = "x" in diccionario
print(f"\n¿Existe 'b'?: {existe_b}")
print(f"¿Existe 'x'?: {existe_x}")

# Obtener colecciones de claves, valores o pares
claves = diccionario.keys()
valores = diccionario.values()
items = diccionario.items()

print("Claves:", claves)
print("Valores:", valores)
print("Pares clave-valor:", items)

# Convertir vistas a listas
lista_claves = list(claves)
lista_valores = list(valores)
lista_items = list(items)

print("Lista de claves:", lista_claves)
print("Lista de valores:", lista_valores)
print("Lista de pares clave-valor:", lista_items)

# Diccionarios anidados
empleados = {
   "E001": {
       "nombre": "Ana García",
       "departamento": "Ingeniería",
       "salario": 50000
   },
   "E002": {
       "nombre": "Carlos López",
       "departamento": "Marketing",
       "salario": 45000
   }
}

# Acceso a datos anidados
nombre_e001 = empleados["E001"]["nombre"]
print("\nNombre del empleado E001:", nombre_e001)

# Modificación de datos anidados
empleados["E002"]["salario"] = 48000
print("Datos actualizados del empleado E002:", empleados["E002"])

# Agregar un nuevo empleado
empleados["E003"] = {
   "nombre": "Laura Martínez",
   "departamento": "Recursos Humanos",
   "salario": 52000
}

# Recorrer el diccionario anidado
print("\nLista de empleados:")
for id_empleado, datos in empleados.items():
   print(f"ID: {id_empleado} - {datos['nombre']} ({datos['departamento']}): ${datos['salario']}")

# ==================================================================================
# 4. SETS (CONJUNTOS) EN PYTHON
# ==================================================================================
print("\n" + "=" * 80)
print("SETS (CONJUNTOS) EN PYTHON")
print("=" * 80)

# Creación de conjuntos
primos = {2, 3, 5, 7, 11, 13}
vocales = {"a", "e", "i", "o", "u"}

# Conjunto vacío (no se puede usar {})
conjunto_vacio = set()

# Creación a partir de iterables
caracteres = set("Hola mundo")

# Los duplicados se eliminan automáticamente
numeros = {1, 2, 2, 3, 3, 3, 4, 4, 4, 4}

print("Números primos:", primos)
print("Vocales:", vocales)
print("Caracteres en 'Hola mundo':", caracteres)
print("Números sin duplicados:", numeros)

# Modificación de conjuntos
conjunto = {1, 2, 3}
print("\nConjunto original:", conjunto)

# Añadir elementos
conjunto.add(4)
print("Después de añadir 4:", conjunto)

# Añadir múltiples elementos
conjunto.update([5, 6, 7])
print("Después de update con [5, 6, 7]:", conjunto)

# Eliminar elementos
conjunto.remove(3)  # Genera error si no existe
print("Después de eliminar 3:", conjunto)

# Eliminar de forma segura
conjunto.discard(10)  # No genera error si no existe
print("Después de discard 10 (que no existía):", conjunto)

# Extraer elemento arbitrario
elemento = conjunto.pop()
print(f"Elemento extraído: {elemento}, Conjunto resultante: {conjunto}")

# Operaciones entre conjuntos
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print("\nA =", A)
print("B =", B)

# Unión (A ∪ B): todos los elementos que están en A o en B
union = A | B
union_alt = A.union(B)  # Método alternativo
print("A ∪ B =", union)
print("A.union(B) =", union_alt)

# Intersección (A ∩ B): elementos comunes a A y B
interseccion = A & B
interseccion_alt = A.intersection(B)
print("A ∩ B =", interseccion)
print("A.intersection(B) =", interseccion_alt)

# Diferencia (A - B): elementos en A que no están en B
diferencia = A - B
diferencia_alt = A.difference(B)
print("A - B =", diferencia)
print("A.difference(B) =", diferencia_alt)

# Diferencia simétrica (A ∆ B): elementos que están en A o B, pero no en ambos
dif_simetrica = A ^ B
dif_simetrica_alt = A.symmetric_difference(B)
print("A ∆ B =", dif_simetrica)
print("A.symmetric_difference(B) =", dif_simetrica_alt)

# Comparación entre conjuntos
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}
C = {6, 7, 8}

# Subconjunto
print("\n¿A es subconjunto de B?", A <= B)  # Comprueba si todos los elementos de A están en B
print("¿A es subconjunto propio de B?", A < B)  # Comprueba si A es subconjunto de B pero A ≠ B
print("A.issubset(B) =", A.issubset(B))  # Método alternativo

# Superconjunto
print("¿B es superconjunto de A?", B >= A)  # Comprueba si todos los elementos de A están en B
print("B.issuperset(A) =", B.issuperset(A))  # Método alternativo

# Conjuntos disjuntos
print("¿A y C son disjuntos?", A.isdisjoint(C))  # Comprueba si A y C no tienen elementos en común

# Conjuntos inmutables (frozenset)
vocales_inmutables = frozenset({"a", "e", "i", "o", "u"})
print("\nConjunto inmutable de vocales:", vocales_inmutables)

# Uso como clave en diccionarios
grupos_foneticos = {
   frozenset({"b", "p", "m"}): "Bilabiales",
   frozenset({"d", "t", "n"}): "Dentales"
}
print("Grupos fonéticos:", grupos_foneticos)

# ==================================================================================
# 5. APLICACIÓN EN APRENDIZAJE POR REFUERZO
# ==================================================================================
print("\n" + "=" * 80)
print("APLICACIÓN EN APRENDIZAJE POR REFUERZO: Q-LEARNING SIMPLIFICADO")
print("=" * 80)

# Implementación de Q-Learning para un entorno de grid world simple
class SimpleGridWorld:
   def __init__(self, width=5, height=5):
       self.width = width
       self.height = height
       self.agent_pos = (0, 0)  # Inicio en la esquina superior izquierda
       self.goal_pos = (width-1, height-1)  # Objetivo en la esquina inferior derecha
       self.obstacles = {(1, 1), (2, 2), (3, 1), (1, 3)}  # Obstáculos como conjunto
       self.actions = {0: (-1, 0),  # Arriba
                     1: (0, 1),   # Derecha
                     2: (1, 0),   # Abajo
                     3: (0, -1)}  # Izquierda

   def reset(self):
       self.agent_pos = (0, 0)
       return self.agent_pos

   def step(self, action):
       # Calcular nueva posición
       dx, dy = self.actions[action]
       new_x = max(0, min(self.width-1, self.agent_pos[0] + dx))
       new_y = max(0, min(self.height-1, self.agent_pos[1] + dy))
       new_pos = (new_x, new_y)

       # Verificar si es obstáculo
       if new_pos in self.obstacles:
           new_pos = self.agent_pos  # No se mueve
           reward = -1
       elif new_pos == self.goal_pos:
           reward = 10  # Llegó a la meta
           done = True
           self.agent_pos = new_pos
           return new_pos, reward, done
       else:
           reward = -0.1  # Pequeña penalización por cada paso

       self.agent_pos = new_pos
       done = False  # No ha terminado el episodio
       return new_pos, reward, done

   def render(self):
       grid = []
       for i in range(self.height):
           row = []
           for j in range(self.width):
               if (i, j) == self.agent_pos:
                   row.append('A')  # Agente
               elif (i, j) == self.goal_pos:
                   row.append('G')  # Meta
               elif (i, j) in self.obstacles:
                   row.append('X')  # Obstáculo
               else:
                   row.append('.')  # Espacio vacío
           grid.append(row)

       # Imprimir la cuadrícula
       for row in grid:
           print(' '.join(row))
       print()

# Algoritmo Q-Learning
def q_learning(env, episodes=100, alpha=0.1, gamma=0.9, epsilon=0.1):
   # Inicializar tabla Q como diccionario de diccionarios
   q_table = {}

   # Almacenar datos para graficar
   rewards_per_episode = []
   steps_per_episode = []

   for episode in range(episodes):
       state = env.reset()  # Obtener estado inicial
       total_reward = 0
       steps = 0
       done = False

       while not done:
           # Exploración vs explotación
           if random.random() < epsilon:
               action = random.choice(list(env.actions.keys()))  # Exploración
           else:
               # Si el estado no está en la tabla Q, inicializarlo
               if state not in q_table:
                   q_table[state] = {a: 0 for a in env.actions}

               # Obtener la acción con mayor valor Q
               action = max(q_table[state], key=q_table[state].get)  # Explotación

           # Ejecutar acción
           next_state, reward, done = env.step(action)
           total_reward += reward
           steps += 1

           # Actualizar tabla Q
           if state not in q_table:
               q_table[state] = {a: 0 for a in env.actions}

           if next_state not in q_table:
               q_table[next_state] = {a: 0 for a in env.actions}

           # Fórmula de Q-learning
           best_next_action = max(q_table[next_state], key=q_table[next_state].get)
           q_table[state][action] = (1 - alpha) * q_table[state][action] + \
                                    alpha * (reward + gamma * q_table[next_state][best_next_action])

           state = next_state

       rewards_per_episode.append(total_reward)
       steps_per_episode.append(steps)

       if (episode + 1) % 10 == 0:
           print(f"Episodio {episode + 1}/{episodes}, Recompensa: {total_reward:.1f}, Pasos: {steps}")

   return q_table, rewards_per_episode, steps_per_episode

# Ejecutar Q-Learning
env = SimpleGridWorld()
print("Entorno inicial:")
env.render()

print("Entrenando agente con Q-Learning...")
q_table, rewards, steps = q_learning(env, episodes=100)

# Evaluar el agente entrenado
print("\nEvaluación del agente entrenado:")
state = env.reset()
env.render()
done = False
total_reward = 0
total_steps = 0

while not done:
   if state not in q_table:
       action = random.choice(list(env.actions.keys()))
   else:
       action = max(q_table[state], key=q_table[state].get)

   next_state, reward, done = env.step(action)
   total_reward += reward
   total_steps += 1

   print(f"Paso {total_steps}: Acción {action}, Recompensa {reward}")
   env.render()

   state = next_state

print(f"Evaluación completada: Recompensa total {total_reward}, Pasos {total_steps}")

# Visualizar el aprendizaje
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(rewards)
plt.title('Recompensa por episodio')
plt.xlabel('Episodio')
plt.ylabel('Recompensa total')

plt.subplot(1, 2, 2)
plt.plot(steps)
plt.title('Pasos por episodio')
plt.xlabel('Episodio')
plt.ylabel('Número de pasos')

plt.tight_layout()
plt.show()

# ==================================================================================
# 6. CONCLUSIONES
# ==================================================================================
print("\n" + "=" * 80)
print("CONCLUSIONES")
print("=" * 80)

print("""
Las estructuras de datos avanzadas en Python proporcionan herramientas poderosas
para implementar algoritmos de Aprendizaje por Refuerzo:

1) LISTAS: Flexibles y mutables, ideales para almacenar secuencias de estados,
  acciones y recompensas durante episodios.

2) TUPLAS: Inmutables y eficientes, perfectas para representar estados y transiciones
  (estado, acción, recompensa, siguiente_estado).

3) DICCIONARIOS: Rápido acceso por clave, excelentes para implementar tablas Q
  y almacenar configuraciones de entornos.

4) SETS: Colecciones de elementos únicos, perfectos para seguimiento de estados
  visitados y eliminación eficiente de duplicados.

Dominar estas estructuras es fundamental para implementar algoritmos efectivos de RL.
""")
```
