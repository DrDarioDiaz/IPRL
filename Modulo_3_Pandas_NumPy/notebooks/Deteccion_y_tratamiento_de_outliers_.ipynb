```python
"""
# Métodos de Detección y Tratamiento de Outliers Univariantes y Multivariantes

Este notebook presenta un estudio completo sobre los principales métodos para detectar y tratar valores atípicos (outliers),
tanto en contextos univariantes como multivariantes.

Autor: Dr. Darío Ezequiel Díaz
"""

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.covariance import EllipticEnvelope
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configurar el estilo de visualización
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Semilla para reproducibilidad
np.random.seed(42)

#-----------------------------------------------------------------
# 1. INTRODUCCIÓN A LOS OUTLIERS
#-----------------------------------------------------------------

def crear_datos_con_outliers():
    """
    Crea conjuntos de datos sintéticos con outliers para demostrar las técnicas
    de detección y tratamiento.
    """
    # Conjunto de datos univariante
    n_samples = 1000

    # Distribución normal para la mayoría de los datos
    x = np.random.normal(0, 1, n_samples)

    # Agregar outliers univariantes
    outliers_indices = np.random.choice(range(n_samples), size=20, replace=False)
    x[outliers_indices] = np.random.uniform(5, 10, 20) * np.random.choice([-1, 1], 20)

    # Conjunto de datos bivariante con outliers
    x1 = np.random.normal(0, 1, n_samples)
    # X2 correlacionado con X1
    x2 = 0.8 * x1 + np.random.normal(0, 0.5, n_samples)

    # Agregar outliers bivariantes
    outliers_indices_biv = np.random.choice(range(n_samples), size=20, replace=False)
    # Algunos outliers extremos en ambas dimensiones
    x1[outliers_indices_biv[:5]] = np.random.uniform(7, 10, 5) * np.random.choice([-1, 1], 5)
    x2[outliers_indices_biv[:5]] = np.random.uniform(7, 10, 5) * np.random.choice([-1, 1], 5)

    # Algunos outliers en una sola dimensión
    x1[outliers_indices_biv[5:10]] = np.random.uniform(7, 10, 5) * np.random.choice([-1, 1], 5)
    x2[outliers_indices_biv[10:15]] = np.random.uniform(7, 10, 5) * np.random.choice([-1, 1], 5)

    # Algunos outliers que rompen la correlación pero no son extremos en ninguna dimensión
    x1[outliers_indices_biv[15:]] = np.random.uniform(2, 4, 5)
    x2[outliers_indices_biv[15:]] = np.random.uniform(-4, -2, 5)

    # Crear conjunto de datos multivariante (más de 2 dimensiones)
    n_features = 5
    X_multi = np.random.normal(0, 1, (n_samples, n_features))

    # Correlación entre características
    for i in range(1, n_features):
        X_multi[:, i] = 0.3 * X_multi[:, 0] + 0.4 * X_multi[:, i-1] + np.random.normal(0, 0.5, n_samples)

    # Agregar outliers multivariantes
    outliers_indices_multi = np.random.choice(range(n_samples), size=30, replace=False)

    # Outliers en múltiples dimensiones
    for i in range(n_features):
        X_multi[outliers_indices_multi[:10], i] = np.random.uniform(5, 10, 10) * np.random.choice([-1, 1], 10)

    # Outliers en dimensiones específicas
    for i in range(n_features):
        subset = outliers_indices_multi[10+i:20+i]
        if len(subset) > 0:  # Asegurarse de que hay suficientes índices
            X_multi[subset, i] = np.random.uniform(5, 10, len(subset)) * np.random.choice([-1, 1], len(subset))

    # Crear DataFrames
    df_uni = pd.DataFrame({'valor': x})
    df_biv = pd.DataFrame({'x1': x1, 'x2': x2})
    df_multi = pd.DataFrame(X_multi, columns=[f'X{i+1}' for i in range(n_features)])

    # Agregar variable categórica para demostrar tratamiento de outliers categóricos
    categorias = ['A', 'B', 'C', 'D']
    probabilidades = [0.7, 0.2, 0.05, 0.05]  # D es raro, potencialmente un outlier
    df_uni['categoria'] = np.random.choice(categorias, size=n_samples, p=probabilidades)

    return df_uni, df_biv, df_multi, outliers_indices, outliers_indices_biv, outliers_indices_multi

# Crear los conjuntos de datos con outliers
df_uni, df_biv, df_multi, true_outliers_uni, true_outliers_biv, true_outliers_multi = crear_datos_con_outliers()

# Visualizar los datos
print("INTRODUCCIÓN A LOS OUTLIERS")
print("--------------------------")
print("Muestra del conjunto de datos univariante:")
print(df_uni.head())

print("\nMuestra del conjunto de datos bivariante:")
print(df_biv.head())

print("\nMuestra del conjunto de datos multivariante:")
print(df_multi.head())

print("\nEstadísticas descriptivas del conjunto univariante:")
print(df_uni['valor'].describe())

# ¿Qué son los outliers?
print("""
¿Qué son los outliers?
---------------------
Los outliers (valores atípicos) son observaciones que se desvían significativamente del patrón
general de los datos. Estos pueden surgir por diversas razones:

1. Errores de medición o registro
2. Fenómenos naturales poco frecuentes pero válidos
3. Comportamientos anómalos en el sistema estudiado
4. Fraudes o actividades inusuales

Los outliers pueden ser:
- Univariantes: Valores extremos en una sola variable
- Multivariantes: Observaciones anómalas considerando múltiples variables simultáneamente

El impacto de los outliers puede ser significativo:
- Distorsionan las medidas de tendencia central y dispersión
- Afectan las correlaciones y relaciones entre variables
- Comprometen la validez de modelos estadísticos y de aprendizaje automático
- Pueden sesgar los resultados y conclusiones del análisis

Por tanto, su correcta identificación y tratamiento es una etapa crucial del preprocesamiento de datos.
""")

# Visualización de outliers univariantes
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.boxplot(df_uni['valor'])
plt.title('Boxplot de datos univariantes')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
sns.histplot(df_uni['valor'], kde=True, bins=50)
plt.title('Histograma de datos univariantes')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Visualización de outliers bivariantes
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.scatter(df_biv['x1'], df_biv['x2'], alpha=0.5)
plt.title('Scatter plot de datos bivariantes')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True, alpha=0.3)

# Marcar los outliers verdaderos
plt.scatter(df_biv.iloc[true_outliers_biv]['x1'], df_biv.iloc[true_outliers_biv]['x2'],
            color='red', alpha=0.7, s=80, edgecolors='k', label='Outliers reales')
plt.legend()

plt.subplot(2, 1, 2)
sns.kdeplot(data=df_biv, x='x1', y='x2', fill=True, cmap='viridis')
plt.title('Densidad bivariante')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

#-----------------------------------------------------------------
# 2. DETECCIÓN DE OUTLIERS UNIVARIANTES
#-----------------------------------------------------------------

print("\nMÉTODOS DE DETECCIÓN DE OUTLIERS UNIVARIANTES")
print("--------------------------------------------")

# 2.1 Método Z-score
def detectar_outliers_zscore(datos, umbral=3.0):
    """
    Detecta outliers utilizando el método Z-score.

    Parámetros:
    -----------
    datos : array-like
        Los datos a analizar
    umbral : float
        El umbral para considerar un valor como outlier (típicamente 3)

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    z_scores = stats.zscore(datos)
    return np.where(np.abs(z_scores) > umbral)[0]

# 2.2 Método Z-score modificado (MAD)
def detectar_outliers_zscore_modificado(datos, umbral=3.5):
    """
    Detecta outliers utilizando el método Z-score modificado basado en MAD.
    Es más robusto frente a la presencia de outliers.

    Parámetros:
    -----------
    datos : array-like
        Los datos a analizar
    umbral : float
        El umbral para considerar un valor como outlier (típicamente 3.5)

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    # Calcular la mediana
    mediana = np.median(datos)
    # Calcular MAD (Median Absolute Deviation)
    mad = np.median(np.abs(datos - mediana))
    # Factor de consistencia para distribución normal
    factor_consistencia = 1.4826  # Este factor hace que MAD sea comparable con desviación estándar
    # Calcular Z-score modificado
    z_mod = factor_consistencia * (datos - mediana) / mad
    return np.where(np.abs(z_mod) > umbral)[0]

# 2.3 Método IQR (Rango Intercuartílico)
def detectar_outliers_iqr(datos, multiplicador=1.5):
    """
    Detecta outliers utilizando el método del rango intercuartílico (IQR).

    Parámetros:
    -----------
    datos : array-like
        Los datos a analizar
    multiplicador : float
        Factor para multiplicar el IQR (típicamente 1.5 para outliers potenciales,
        3.0 para outliers extremos)

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    q1 = np.percentile(datos, 25)
    q3 = np.percentile(datos, 75)
    iqr = q3 - q1
    lower_bound = q1 - multiplicador * iqr
    upper_bound = q3 + multiplicador * iqr
    return np.where((datos < lower_bound) | (datos > upper_bound))[0]

# 2.4 Método percentiles
def detectar_outliers_percentiles(datos, percentil_inferior=1, percentil_superior=99):
    """
    Detecta outliers utilizando límites basados en percentiles.

    Parámetros:
    -----------
    datos : array-like
        Los datos a analizar
    percentil_inferior : float
        Percentil inferior (típicamente entre 0.1 y 5)
    percentil_superior : float
        Percentil superior (típicamente entre 95 y 99.9)

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    limit_inf = np.percentile(datos, percentil_inferior)
    limit_sup = np.percentile(datos, percentil_superior)
    return np.where((datos < limit_inf) | (datos > limit_sup))[0]

# 2.5 DBSCAN para detección univariante
def detectar_outliers_dbscan(datos, eps=0.5, min_samples=5):
    """
    Detecta outliers utilizando DBSCAN aplicado a datos univariantes.

    Parámetros:
    -----------
    datos : array-like
        Los datos a analizar
    eps : float
        El radio máximo del vecindario
    min_samples : int
        Número mínimo de muestras en un vecindario

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    X = datos.reshape(-1, 1)  # Reshape para que DBSCAN lo acepte
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
    # DBSCAN asigna -1 a los outliers
    return np.where(clustering.labels_ == -1)[0]

# Aplicar los métodos a los datos univariantes
datos_uni = df_uni['valor'].values

outliers_zscore = detectar_outliers_zscore(datos_uni)
outliers_zscore_mod = detectar_outliers_zscore_modificado(datos_uni)
outliers_iqr = detectar_outliers_iqr(datos_uni)
outliers_percentiles = detectar_outliers_percentiles(datos_uni)
outliers_dbscan = detectar_outliers_dbscan(datos_uni, eps=0.5, min_samples=5)

# Crear un DataFrame para comparar resultados con los outliers reales
df_comparacion = pd.DataFrame({
    'Índice': range(len(datos_uni)),
    'Valor': datos_uni,
    'Es outlier real': np.isin(np.arange(len(datos_uni)), true_outliers_uni),
    'Detectado por Z-score': np.isin(np.arange(len(datos_uni)), outliers_zscore),
    'Detectado por Z-score mod': np.isin(np.arange(len(datos_uni)), outliers_zscore_mod),
    'Detectado por IQR': np.isin(np.arange(len(datos_uni)), outliers_iqr),
    'Detectado por Percentiles': np.isin(np.arange(len(datos_uni)), outliers_percentiles),
    'Detectado por DBSCAN': np.isin(np.arange(len(datos_uni)), outliers_dbscan)
})

# Mostrar los resultados de la detección
print("\nResultados de la detección de outliers univariantes:")
print(f"Número de outliers reales: {sum(df_comparacion['Es outlier real'])}")
print(f"Detectados por Z-score: {sum(df_comparacion['Detectado por Z-score'])}")
print(f"Detectados por Z-score modificado: {sum(df_comparacion['Detectado por Z-score mod'])}")
print(f"Detectados por IQR: {sum(df_comparacion['Detectado por IQR'])}")
print(f"Detectados por Percentiles: {sum(df_comparacion['Detectado por Percentiles'])}")
print(f"Detectados por DBSCAN: {sum(df_comparacion['Detectado por DBSCAN'])}")

print("\nVerdaderos positivos (outliers correctamente identificados):")
print(f"Z-score: {sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score'])}")
print(f"Z-score mod: {sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score mod'])}")
print(f"IQR: {sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por IQR'])}")
print(f"Percentiles: {sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Percentiles'])}")
print(f"DBSCAN: {sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por DBSCAN'])}")

# Visualizar los resultados
plt.figure(figsize=(15, 10))

for i, (metodo, outliers) in enumerate([
    ('Z-score', outliers_zscore),
    ('Z-score modificado', outliers_zscore_mod),
    ('IQR', outliers_iqr),
    ('Percentiles', outliers_percentiles),
    ('DBSCAN', outliers_dbscan)
]):
    plt.subplot(3, 2, i+1)
    plt.hist(datos_uni, bins=50, alpha=0.5, label='Datos')

    if len(outliers) > 0:
        plt.hist(datos_uni[outliers], bins=50, color='red', alpha=0.8, label='Outliers detectados')

    plt.axvline(np.mean(datos_uni), color='green', linestyle='--', label='Media')
    plt.axvline(np.median(datos_uni), color='blue', linestyle='--', label='Mediana')

    plt.title(f'Detección con {metodo}')
    plt.legend()

plt.subplot(3, 2, 6)
plt.hist(datos_uni, bins=50, alpha=0.5, label='Datos')
plt.hist(datos_uni[true_outliers_uni], bins=50, color='red', alpha=0.8, label='Outliers reales')
plt.axvline(np.mean(datos_uni), color='green', linestyle='--', label='Media')
plt.axvline(np.median(datos_uni), color='blue', linestyle='--', label='Mediana')
plt.title('Outliers reales (generados)')
plt.legend()

plt.tight_layout()
plt.show()

# Comparación de las técnicas
info_metodos = pd.DataFrame({
    'Método': ['Z-score', 'Z-score Modificado', 'IQR', 'Percentiles', 'DBSCAN'],
    'Detectados': [
        len(outliers_zscore),
        len(outliers_zscore_mod),
        len(outliers_iqr),
        len(outliers_percentiles),
        len(outliers_dbscan)
    ],
    'Verdaderos Positivos': [
        sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score']),
        sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score mod']),
        sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por IQR']),
        sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por Percentiles']),
        sum(df_comparacion['Es outlier real'] & df_comparacion['Detectado por DBSCAN'])
    ],
    'Falsos Positivos': [
        sum(~df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score']),
        sum(~df_comparacion['Es outlier real'] & df_comparacion['Detectado por Z-score mod']),
        sum(~df_comparacion['Es outlier real'] & df_comparacion['Detectado por IQR']),
        sum(~df_comparacion['Es outlier real'] & df_comparacion['Detectado por Percentiles']),
        sum(~df_comparacion['Es outlier real'] & df_comparacion['Detectado por DBSCAN'])
    ],
    'Falsos Negativos': [
        sum(df_comparacion['Es outlier real'] & ~df_comparacion['Detectado por Z-score']),
        sum(df_comparacion['Es outlier real'] & ~df_comparacion['Detectado por Z-score mod']),
        sum(df_comparacion['Es outlier real'] & ~df_comparacion['Detectado por IQR']),
        sum(df_comparacion['Es outlier real'] & ~df_comparacion['Detectado por Percentiles']),
        sum(df_comparacion['Es outlier real'] & ~df_comparacion['Detectado por DBSCAN'])
    ]
})

# Agregar precisión y recall
info_metodos['Precisión'] = info_metodos['Verdaderos Positivos'] / (info_metodos['Verdaderos Positivos'] + info_metodos['Falsos Positivos'])
info_metodos['Recall'] = info_metodos['Verdaderos Positivos'] / (info_metodos['Verdaderos Positivos'] + info_metodos['Falsos Negativos'])
info_metodos['F1-Score'] = 2 * (info_metodos['Precisión'] * info_metodos['Recall']) / (info_metodos['Precisión'] + info_metodos['Recall'])

print("\nEvaluación de métodos de detección de outliers univariantes:")
print(info_metodos)

# Visualización comparativa de métricas
plt.figure(figsize=(12, 6))

# Gráfico de barras para precisión y recall
plt.subplot(1, 2, 1)
info_metodos.plot(x='Método', y=['Precisión', 'Recall', 'F1-Score'], kind='bar', ax=plt.gca())
plt.title('Métricas de evaluación')
plt.ylabel('Valor')
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)

# Gráfico de barras para conteos
plt.subplot(1, 2, 2)
info_metodos.plot(x='Método', y=['Verdaderos Positivos', 'Falsos Positivos', 'Falsos Negativos'], kind='bar', ax=plt.gca())
plt.title('Conteo de tipos de clasificación')
plt.ylabel('Número de observaciones')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("""
Conclusiones sobre métodos univariantes:
----------------------------------------
1. Z-score estándar es sensible a outliers en el cálculo de media y desviación estándar.
2. Z-score modificado (basado en MAD) es más robusto cuando hay múltiples outliers.
3. IQR es un método robusto y ampliamente utilizado, especialmente eficaz para distribuciones sesgadas.
4. Los percentiles son flexibles pero requieren ajustes específicos según el contexto.
5. DBSCAN puede ser efectivo para detectar clusters de outliers, pero requiere ajuste cuidadoso de parámetros.

La elección del método dependerá de:
- Distribución de los datos
- Proporción esperada de outliers
- Impacto de falsos positivos vs. falsos negativos
- Necesidad de robustez
""")

# 2.6 Outliers categóricos
def detectar_outliers_categoricos(datos, umbral_frecuencia=0.05):
    """
    Detecta outliers en variables categóricas basándose en la frecuencia relativa.

    Parámetros:
    -----------
    datos : Series
        Los datos categóricos a analizar
    umbral_frecuencia : float
        Umbral de frecuencia relativa por debajo del cual una categoría
        se considera un outlier (ej: 0.05 = 5%)

    Retorna:
    --------
    categorias_outliers : list
        Lista de categorías consideradas outliers
    indices : array
        Índices de las observaciones con categorías outliers
    """
    # Calcular frecuencias relativas
    frecuencias = datos.value_counts(normalize=True)
    # Identificar categorías raras
    categorias_outliers = frecuencias[frecuencias < umbral_frecuencia].index.tolist()
    # Encontrar observaciones con esas categorías
    indices = datos[datos.isin(categorias_outliers)].index.values

    return categorias_outliers, indices

# Detectar outliers categóricos
categorias_outliers, indices_cat_outliers = detectar_outliers_categoricos(df_uni['categoria'])

print("\nDetección de outliers categóricos:")
print(f"Categorías consideradas outliers: {categorias_outliers}")
print(f"Número de observaciones con categorías outliers: {len(indices_cat_outliers)}")

# Visualizar las frecuencias de las categorías
plt.figure(figsize=(10, 6))
sns.countplot(x='categoria', data=df_uni, order=df_uni['categoria'].value_counts().index)
plt.title('Frecuencia de categorías')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')

# Marcar las categorías outliers
for categoria in categorias_outliers:
    plt.text(list(df_uni['categoria'].value_counts().index).index(categoria),
             df_uni['categoria'].value_counts()[categoria] + 5,
             'OUTLIER', color='red', ha='center')

plt.grid(True, alpha=0.3)
plt.show()

print("""
Outliers categóricos:
--------------------
Los outliers categóricos son categorías que aparecen con una frecuencia inusualmente baja.
Posibles causas:
1. Errores de codificación o entrada de datos
2. Categorías legítimas pero raras
3. Fusión de datos de diferentes fuentes con categorías incompatibles

Métodos para detectarlos:
1. Análisis de frecuencia relativa (mostrado arriba)
2. Reglas de negocio o dominio específico
3. Análisis de correspondencia múltiple para variables categóricas múltiples

Estrategias de tratamiento:
1. Agrupar categorías raras en una categoría "Otros"
2. Eliminar observaciones con categorías raras si son pocas
3. Mantenerlas si son importantes a pesar de su baja frecuencia
4. Aplicar técnicas de codificación robustas (como target encoding)
""")

#-----------------------------------------------------------------
# 3. DETECCIÓN DE OUTLIERS MULTIVARIANTES
#-----------------------------------------------------------------

print("\nMÉTODOS DE DETECCIÓN DE OUTLIERS MULTIVARIANTES")
print("------------------------------------------------")

# 3.1 Distancia de Mahalanobis
def detectar_outliers_mahalanobis(datos, umbral_chi2=0.99):
    """
    Detecta outliers multivariantes utilizando la distancia de Mahalanobis.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    umbral_chi2 : float
        Percentil de la distribución chi-cuadrado para el umbral (típicamente 0.95 o 0.99)

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    distancias : array
        Distancias de Mahalanobis para cada observación
    """
    # Calcular media y covarianza
    media = np.mean(datos, axis=0)
    cov = np.cov(datos, rowvar=False)

    # Invertir matriz de covarianza
    try:
        inv_cov = np.linalg.inv(cov)
    except np.linalg.LinAlgError:
        # Si la matriz no es invertible, usar pseudoinversa
        inv_cov = np.linalg.pinv(cov)

    # Calcular distancias de Mahalanobis
    distancias = []
    for i in range(datos.shape[0]):
        x = datos[i]
        # Fórmula de distancia de Mahalanobis
        dist = np.sqrt(np.dot(np.dot((x - media), inv_cov), (x - media).T))
        distancias.append(dist)

    distancias = np.array(distancias)

    # Umbral basado en distribución chi-cuadrado
    # La distancia de Mahalanobis al cuadrado sigue una distribución chi-cuadrado
    # con grados de libertad igual al número de variables
    grados_libertad = datos.shape[1]
    umbral = stats.chi2.ppf(umbral_chi2, grados_libertad)

    # Identificar outliers
    outliers = np.where(distancias**2 > umbral)[0]

    return outliers, distancias

# 3.2 Método basado en PCA
def detectar_outliers_pca(datos, n_componentes=2, umbral_std=3):
    """
    Detecta outliers multivariantes utilizando PCA y Z-scores sobre los componentes principales.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    n_componentes : int
        Número de componentes principales a utilizar
    umbral_std : float
        Número de desviaciones estándar para considerar un punto como outlier

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    scores : array
        Puntuaciones de los componentes principales
    """
    # Aplicar PCA
    pca = PCA(n_components=n_componentes)
    scores = pca.fit_transform(datos)

    # Calcular Z-scores para cada componente
    z_scores = np.zeros_like(scores)
    for i in range(n_componentes):
        z_scores[:, i] = (scores[:, i] - np.mean(scores[:, i])) / np.std(scores[:, i])

    # Calcular la magnitud del Z-score (distancia euclidiana desde el origen en el espacio de Z-scores)
    z_magnitudes = np.sqrt(np.sum(z_scores**2, axis=1))

    # Identificar outliers
    outliers = np.where(z_magnitudes > umbral_std)[0]

    return outliers, scores

# 3.3 Isolation Forest
def detectar_outliers_isolation_forest(datos, contaminacion=0.05):
    """
    Detecta outliers multivariantes utilizando Isolation Forest.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    contaminacion : float
        Proporción esperada de outliers en los datos

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    model = IsolationForest(contamination=contaminacion, random_state=42)
    model.fit(datos)
    # Las predicciones son 1 para inliers y -1 para outliers
    predicciones = model.predict(datos)
    outliers = np.where(predicciones == -1)[0]
    return outliers

# 3.4 Local Outlier Factor (LOF)
def detectar_outliers_lof(datos, n_vecinos=20, contaminacion=0.05):
    """
    Detecta outliers multivariantes utilizando Local Outlier Factor.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    n_vecinos : int
        Número de vecinos a considerar
    contaminacion : float
        Proporción esperada de outliers en los datos

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    model = LocalOutlierFactor(n_neighbors=n_vecinos, contamination=contaminacion)
    # LOF en modo novelty detection es para predecir después de ajustar
    # En este caso, solo queremos identificar los outliers en los datos
    predicciones = model.fit_predict(datos)
    outliers = np.where(predicciones == -1)[0]
    return outliers

# 3.5 Envelope elíptico
def detectar_outliers_envelope(datos, contaminacion=0.05):
    """
    Detecta outliers multivariantes utilizando el método de Envelope elíptico.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    contaminacion : float
        Proporción esperada de outliers en los datos

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    model = EllipticEnvelope(contamination=contaminacion, random_state=42)
    # Las predicciones son 1 para inliers y -1 para outliers
    model.fit(datos)
    predicciones = model.predict(datos)
    outliers = np.where(predicciones == -1)[0]
    return outliers

# 3.6 DBSCAN para detección multivariante
def detectar_outliers_dbscan_multi(datos, eps=0.5, min_samples=5):
    """
    Detecta outliers multivariantes utilizando DBSCAN.

    Parámetros:
    -----------
    datos : DataFrame o array
        Los datos multivariantes a analizar
    eps : float
        El radio máximo del vecindario
    min_samples : int
        Número mínimo de muestras en un vecindario

    Retorna:
    --------
    indices : array
        Índices de las observaciones consideradas outliers
    """
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(datos)
    # DBSCAN asigna -1 a los outliers
    outliers = np.where(clustering.labels_ == -1)[0]
    return outliers

# Aplicar los métodos a los datos bivariantes
datos_biv = df_biv.values  # Convertir a array numpy
datos_multi = df_multi.values

# Es importante estandarizar los datos para métodos basados en distancia
scaler = StandardScaler()
datos_biv_scaled = scaler.fit_transform(datos_biv)
datos_multi_scaled = scaler.fit_transform(datos_multi)

# Detectar outliers multivariantes con los diferentes métodos
outliers_mahalanobis, distancias_mah = detectar_outliers_mahalanobis(datos_biv_scaled)
outliers_pca, scores_pca = detectar_outliers_pca(datos_biv_scaled)
outliers_iforest = detectar_outliers_isolation_forest(datos_biv_scaled)
outliers_lof = detectar_outliers_lof(datos_biv_scaled)
outliers_envelope = detectar_outliers_envelope(datos_biv_scaled)
outliers_dbscan_multi = detectar_outliers_dbscan_multi(datos_biv_scaled, eps=0.5, min_samples=5)

# Mostrar resultados
print("\nResultados de la detección de outliers multivariantes (datos bivariantes):")
print(f"Número de outliers reales: {len(true_outliers_biv)}")
print(f"Detectados por Mahalanobis: {len(outliers_mahalanobis)}")
print(f"Detectados por PCA: {len(outliers_pca)}")
print(f"Detectados por Isolation Forest: {len(outliers_iforest)}")
print(f"Detectados por LOF: {len(outliers_lof)}")
print(f"Detectados por Envelope elíptico: {len(outliers_envelope)}")
print(f"Detectados por DBSCAN: {len(outliers_dbscan_multi)}")

# Crear un DataFrame para la comparación
df_comp_multi = pd.DataFrame({
    'Índice': range(len(datos_biv)),
    'Es outlier real': np.isin(np.arange(len(datos_biv)), true_outliers_biv),
    'Detectado por Mahalanobis': np.isin(np.arange(len(datos_biv)), outliers_mahalanobis),
    'Detectado por PCA': np.isin(np.arange(len(datos_biv)), outliers_pca),
    'Detectado por Isolation Forest': np.isin(np.arange(len(datos_biv)), outliers_iforest),
    'Detectado por LOF': np.isin(np.arange(len(datos_biv)), outliers_lof),
    'Detectado por Envelope': np.isin(np.arange(len(datos_biv)), outliers_envelope),
    'Detectado por DBSCAN': np.isin(np.arange(len(datos_biv)), outliers_dbscan_multi)
})

# Calcular métricas de rendimiento
print("\nVerdaderos positivos (outliers correctamente identificados):")
print(f"Mahalanobis: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Mahalanobis'])}")
print(f"PCA: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por PCA'])}")
print(f"Isolation Forest: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Isolation Forest'])}")
print(f"LOF: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por LOF'])}")
print(f"Envelope: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Envelope'])}")
print(f"DBSCAN: {sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por DBSCAN'])}")

# Visualizar los resultados
plt.figure(figsize=(15, 10))

# Función para visualizar los outliers detectados en gráficos de dispersión
def plot_outliers_scatter(ax, x, y, outliers, title):
    ax.scatter(x, y, alpha=0.5, label='Datos')
    if len(outliers) > 0:
        ax.scatter(x[outliers], y[outliers], color='red', alpha=0.8, label='Outliers detectados')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)

# Visualizar para cada método
for i, (metodo, outliers) in enumerate([
    ('Mahalanobis', outliers_mahalanobis),
    ('PCA', outliers_pca),
    ('Isolation Forest', outliers_iforest),
    ('LOF', outliers_lof),
    ('Envelope elíptico', outliers_envelope),
    ('DBSCAN', outliers_dbscan_multi)
]):
    ax = plt.subplot(3, 2, i+1)
    plot_outliers_scatter(ax, df_biv['x1'], df_biv['x2'], outliers, f'Detección con {metodo}')

plt.tight_layout()
plt.show()

# Visualización de métricas específicas para cada método
plt.figure(figsize=(15, 10))

# Distancia de Mahalanobis
plt.subplot(2, 2, 1)
plt.hist(distancias_mah, bins=50, alpha=0.7)
if len(outliers_mahalanobis) > 0:
    plt.axvline(distancias_mah[outliers_mahalanobis].min(), color='red', linestyle='--',
                label=f'Umbral de detección: {distancias_mah[outliers_mahalanobis].min():.2f}')
plt.title('Distribución de distancias de Mahalanobis')
plt.xlabel('Distancia')
plt.ylabel('Frecuencia')
plt.legend()

# PCA
plt.subplot(2, 2, 2)
plt.scatter(scores_pca[:, 0], scores_pca[:, 1], alpha=0.5)
if len(outliers_pca) > 0:
    plt.scatter(scores_pca[outliers_pca, 0], scores_pca[outliers_pca, 1], color='red', alpha=0.8)
plt.title('Puntuaciones PCA (primeros 2 componentes)')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Isolation Forest
plt.subplot(2, 2, 3)
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(datos_biv_scaled)
scores = -model.score_samples(datos_biv_scaled)  # Scores negativos (mayor = más anómalo)
plt.hist(scores, bins=50, alpha=0.7)
if len(outliers_iforest) > 0:
    plt.axvline(min(scores[outliers_iforest]), color='red', linestyle='--',
                label=f'Umbral: {min(scores[outliers_iforest]):.2f}')
plt.title('Puntuaciones de anomalía con Isolation Forest')
plt.xlabel('Puntuación de anomalía')
plt.ylabel('Frecuencia')
plt.legend()

# LOF
plt.subplot(2, 2, 4)
model = LocalOutlierFactor(n_neighbors=20, contamination=0.05, novelty=True)
model.fit(datos_biv_scaled)
scores = -model.negative_outlier_factor_  # LOF scores (mayor = más anómalo)
plt.hist(scores, bins=50, alpha=0.7)
if len(outliers_lof) > 0:
    plt.axvline(min(scores[outliers_lof]), color='red', linestyle='--',
                label=f'Umbral: {min(scores[outliers_lof]):.2f}')
plt.title('Factores de outlier local (LOF)')
plt.xlabel('Factor de outlier')
plt.ylabel('Frecuencia')
plt.legend()

plt.tight_layout()
plt.show()

# Comparación de las técnicas multivariantes
info_metodos_multi = pd.DataFrame({
    'Método': ['Mahalanobis', 'PCA', 'Isolation Forest', 'LOF', 'Envelope', 'DBSCAN'],
    'Detectados': [
        len(outliers_mahalanobis),
        len(outliers_pca),
        len(outliers_iforest),
        len(outliers_lof),
        len(outliers_envelope),
        len(outliers_dbscan_multi)
    ],
    'Verdaderos Positivos': [
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Mahalanobis']),
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por PCA']),
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Isolation Forest']),
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por LOF']),
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Envelope']),
        sum(df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por DBSCAN'])
    ],
    'Falsos Positivos': [
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Mahalanobis']),
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por PCA']),
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Isolation Forest']),
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por LOF']),
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por Envelope']),
        sum(~df_comp_multi['Es outlier real'] & df_comp_multi['Detectado por DBSCAN'])
    ],
    'Falsos Negativos': [
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por Mahalanobis']),
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por PCA']),
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por Isolation Forest']),
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por LOF']),
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por Envelope']),
        sum(df_comp_multi['Es outlier real'] & ~df_comp_multi['Detectado por DBSCAN'])
    ]
})

# Agregar precisión y recall
info_metodos_multi['Precisión'] = info_metodos_multi['Verdaderos Positivos'] / (info_metodos_multi['Verdaderos Positivos'] + info_metodos_multi['Falsos Positivos'])
info_metodos_multi['Recall'] = info_metodos_multi['Verdaderos Positivos'] / (info_metodos_multi['Verdaderos Positivos'] + info_metodos_multi['Falsos Negativos'])
info_metodos_multi['F1-Score'] = 2 * (info_metodos_multi['Precisión'] * info_metodos_multi['Recall']) / (info_metodos_multi['Precisión'] + info_metodos_multi['Recall'])

print("\nEvaluación de métodos de detección de outliers multivariantes:")
print(info_metodos_multi)

# Visualización comparativa de métricas
plt.figure(figsize=(12, 6))

# Gráfico de barras para precisión y recall
plt.subplot(1, 2, 1)
info_metodos_multi.plot(x='Método', y=['Precisión', 'Recall', 'F1-Score'], kind='bar', ax=plt.gca())
plt.title('Métricas de evaluación')
plt.ylabel('Valor')
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)

# Gráfico de barras para conteos
plt.subplot(1, 2, 2)
info_metodos_multi.plot(x='Método', y=['Verdaderos Positivos', 'Falsos Positivos', 'Falsos Negativos'], kind='bar', ax=plt.gca())
plt.title('Conteo de tipos de clasificación')
plt.ylabel('Número de observaciones')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("""
Conclusiones sobre métodos multivariantes:
-----------------------------------------
1. La distancia de Mahalanobis es efectiva para datos con distribución aproximadamente normal
   y captura correlaciones entre variables, pero es sensible a la presencia de múltiples outliers.

2. PCA es útil para reducir dimensionalidad y detectar outliers en el subespacio de componentes principales,
   pero puede ocultar outliers en componentes de menor varianza.

3. Isolation Forest es especialmente eficiente en alta dimensionalidad, funciona bien incluso con
   proporción elevada de outliers y no asume ninguna distribución específica.

4. LOF es excelente para detectar outliers locales (puntos que son anómalos respecto a su vecindario)
   pero sensible a la elección del número de vecinos.

5. El Envelope Elíptico asume una distribución normal multivariante y es más eficaz cuando esta suposición
   se cumple razonablemente.

6. DBSCAN identifica bien outliers como puntos aislados, pero es muy sensible a la elección de parámetros
   y puede tener dificultades con datos de densidad variable.

La elección entre estos métodos debería considerar:
- La estructura conocida o esperada de los datos
- El tipo de outliers que se buscan (globales vs. locales)
- La dimensionalidad de los datos
- Los requisitos computacionales
""")

# Demostración en datos de mayor dimensionalidad
# Aquí aplicamos algunos métodos a los datos multidimensionales (5 variables)

print("\nDemostración en datos multidimensionales (5 variables):")

# Aplicar algunos métodos seleccionados
outliers_iforest_multi = detectar_outliers_isolation_forest(datos_multi_scaled)
outliers_lof_multi = detectar_outliers_lof(datos_multi_scaled)
outliers_mahalanobis_multi, _ = detectar_outliers_mahalanobis(datos_multi_scaled)

print(f"Outliers reales: {len(true_outliers_multi)}")
print(f"Detectados por Isolation Forest: {len(outliers_iforest_multi)}")
print(f"Detectados por LOF: {len(outliers_lof_multi)}")
print(f"Detectados por Mahalanobis: {len(outliers_mahalanobis_multi)}")

# Visualización con PCA para datos multidimensionales
pca = PCA(n_components=2)
datos_multi_pca = pca.fit_transform(datos_multi_scaled)

plt.figure(figsize=(15, 5))

# Visualizar datos proyectados en las primeras 2 componentes principales
plt.subplot(1, 3, 1)
plt.scatter(datos_multi_pca[:, 0], datos_multi_pca[:, 1], alpha=0.5, label='Datos')
plt.scatter(datos_multi_pca[true_outliers_multi, 0], datos_multi_pca[true_outliers_multi, 1],
            color='red', alpha=0.8, s=50, label='Outliers reales')
plt.title('Outliers reales (proyección PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True, alpha=0.3)

# Isolation Forest
plt.subplot(1, 3, 2)
plt.scatter(datos_multi_pca[:, 0], datos_multi_pca[:, 1], alpha=0.5, label='Datos')
plt.scatter(datos_multi_pca[outliers_iforest_multi, 0], datos_multi_pca[outliers_iforest_multi, 1],
            color='red', alpha=0.8, s=50, label='Outliers detectados')
plt.title('Isolation Forest (proyección PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True, alpha=0.3)

# LOF
plt.subplot(1, 3, 3)
plt.scatter(datos_multi_pca[:, 0], datos_multi_pca[:, 1], alpha=0.5, label='Datos')
plt.scatter(datos_multi_pca[outliers_lof_multi, 0], datos_multi_pca[outliers_lof_multi, 1],
            color='red', alpha=0.8, s=50, label='Outliers detectados')
plt.title('LOF (proyección PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

#-----------------------------------------------------------------
# 4. TRATAMIENTO DE OUTLIERS
#-----------------------------------------------------------------

print("\nMÉTODOS DE TRATAMIENTO DE OUTLIERS")
print("--------------------------------")

# Utilizaremos los datos univariantes y los outliers detectados con el método IQR
datos_para_tratamiento = df_uni['valor'].copy()
outliers_indices = outliers_iqr  # Utilizar los outliers detectados por IQR

print(f"Número de outliers a tratar: {len(outliers_indices)}")
print(f"Porcentaje de outliers: {len(outliers_indices) / len(datos_para_tratamiento) * 100:.2f}%")

# 4.1 Eliminación de outliers
def tratar_outliers_eliminacion(datos, indices_outliers):
    """
    Elimina los outliers del conjunto de datos.

    Parámetros:
    -----------
    datos : Series o array
        Los datos a tratar
    indices_outliers : array
        Índices de las observaciones consideradas outliers

    Retorna:
    --------
    datos_tratados : Series o array
        Datos sin los outliers
    """
    # Crear una copia para evitar modificar los datos originales
    if isinstance(datos, pd.Series):
        mask = ~datos.index.isin(indices_outliers)
        return datos[mask]
    else:
        mask = ~np.isin(np.arange(len(datos)), indices_outliers)
        return datos[mask]

# 4.2 Imputación (reemplazo) de outliers
def tratar_outliers_imputacion(datos, indices_outliers, metodo='mean'):
    """
    Reemplaza los outliers con un valor estadístico.

    Parámetros:
    -----------
    datos : Series o array
        Los datos a tratar
    indices_outliers : array
        Índices de las observaciones consideradas outliers
    metodo : str
        Método de imputación: 'mean', 'median', 'mode'

    Retorna:
    --------
    datos_tratados : Series o array
        Datos con los outliers reemplazados
    """
    datos_tratados = datos.copy()

    if metodo == 'mean':
        # Calcular la media excluyendo los outliers
        valor_reemplazo = datos[~np.isin(np.arange(len(datos)), indices_outliers)].mean()
    elif metodo == 'median':
        # Calcular la mediana excluyendo los outliers
        valor_reemplazo = datos[~np.isin(np.arange(len(datos)), indices_outliers)].median()
    elif metodo == 'mode':
        # Calcular la moda excluyendo los outliers
        valor_reemplazo = datos[~np.isin(np.arange(len(datos)), indices_outliers)].mode()[0]
    else:
        raise ValueError("Método no válido. Use 'mean', 'median' o 'mode'.")

    # Reemplazar los outliers
    datos_tratados[indices_outliers] = valor_reemplazo

    return datos_tratados

# 4.3 Winsorización (recorte/capping)
def tratar_outliers_winsor(datos, indices_outliers, method='iqr'):
    """
    Reemplaza los outliers con los límites superior e inferior (winsorización).

    Parámetros:
    -----------
    datos : Series o array
        Los datos a tratar
    indices_outliers : array
        Índices de las observaciones consideradas outliers
    method : str
        Método para determinar los límites: 'iqr', 'percentile', 'zscore'

    Retorna:
    --------
    datos_tratados : Series o array
        Datos con los outliers winsorizados
    """
    datos_tratados = datos.copy()

    if method == 'iqr':
        q1 = np.percentile(datos, 25)
        q3 = np.percentile(datos, 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
    elif method == 'percentile':
        lower_bound = np.percentile(datos, 5)
        upper_bound = np.percentile(datos, 95)
    elif method == 'zscore':
        media = np.mean(datos)
        std = np.std(datos)
        lower_bound = media - 3 * std
        upper_bound = media + 3 * std
    else:
        raise ValueError("Método no válido. Use 'iqr', 'percentile' o 'zscore'.")

    # Reemplazar los outliers
    for idx in indices_outliers:
        if datos[idx] < lower_bound:
            datos_tratados[idx] = lower_bound
        elif datos[idx] > upper_bound:
            datos_tratados[idx] = upper_bound

    return datos_tratados, lower_bound, upper_bound

# 4.4 Transformación de variables
def tratar_outliers_transformacion(datos, indices_outliers=None, metodo='log'):
    """
    Aplica una transformación a los datos para reducir el impacto de los outliers.

    Parámetros:
    -----------
    datos : Series o array
        Los datos a tratar
    indices_outliers : array, opcional
        Índices de las observaciones consideradas outliers (no se usa directamente,
        pero se incluye para mantener la firma consistent)
    metodo : str
        Método de transformación: 'log', 'sqrt', 'boxcox'

    Retorna:
    --------
    datos_transformados : Series o array
        Datos transformados
    """
    datos_tratados = datos.copy()

    if metodo == 'log':
        # Log transformation (desplazamiento +1 para manejar valores negativos o cero)
        min_val = datos.min()
        if min_val <= 0:
            datos_tratados = np.log(datos - min_val + 1)
        else:
            datos_tratados = np.log(datos)
    elif metodo == 'sqrt':
        # Square root transformation (desplazamiento para manejar valores negativos)
        min_val = datos.min()
        if min_val < 0:
            datos_tratados = np.sqrt(datos - min_val)
        else:
            datos_tratados = np.sqrt(datos)
    elif metodo == 'boxcox':
        # Box-Cox transformation
        # Requiere datos estrictamente positivos
        min_val = datos.min()
        if min_val <= 0:
            datos_desplazados = datos - min_val + 1
        else:
            datos_desplazados = datos

        # Encontrar lambda óptimo
        datos_tratados, lambda_valor = stats.boxcox(datos_desplazados)
        print(f"Lambda óptimo para Box-Cox: {lambda_valor:.4f}")
    else:
        raise ValueError("Método no válido. Use 'log', 'sqrt' o 'boxcox'.")

    return datos_tratados

# 4.5 Uso de escaladores robustos
def tratar_outliers_robusto(datos, indices_outliers=None):
    """
    Aplica escalado robusto a los datos para reducir el impacto de los outliers.

    Parámetros:
    -----------
    datos : Series o array
        Los datos a tratar
    indices_outliers : array, opcional
        Índices de las observaciones consideradas outliers (no se usa directamente)

    Retorna:
    --------
    datos_escalados : Series o array
        Datos escalados robustamente
    """
    # Garantizar la forma correcta para el escalador
    X = datos.values.reshape(-1, 1) if isinstance(datos, pd.Series) else datos.reshape(-1, 1)

    # Aplicar escalado robusto
    robust_scaler = RobustScaler()
    datos_escalados = robust_scaler.fit_transform(X).flatten()

    return datos_escalados

# Aplicar diferentes métodos de tratamiento
datos_eliminados = tratar_outliers_eliminacion(datos_para_tratamiento, outliers_indices)
datos_imputados_media = tratar_outliers_imputacion(datos_para_tratamiento, outliers_indices, metodo='mean')
datos_imputados_mediana = tratar_outliers_imputacion(datos_para_tratamiento, outliers_indices, metodo='median')
datos_winsor, lower_bound, upper_bound = tratar_outliers_winsor(datos_para_tratamiento, outliers_indices)
datos_log = tratar_outliers_transformacion(datos_para_tratamiento, outliers_indices, metodo='log')
datos_sqrt = tratar_outliers_transformacion(datos_para_tratamiento, outliers_indices, metodo='sqrt')
datos_boxcox = tratar_outliers_transformacion(datos_para_tratamiento, outliers_indices, metodo='boxcox')
datos_robustos = tratar_outliers_robusto(datos_para_tratamiento, outliers_indices)

# Comparar estadísticas descriptivas
stats_comparacion = pd.DataFrame({
    'Original': datos_para_tratamiento.describe(),
    'Eliminación': datos_eliminados.describe(),
    'Imputación (media)': pd.Series(datos_imputados_media).describe(),
    'Imputación (mediana)': pd.Series(datos_imputados_mediana).describe(),
    'Winsorización': pd.Series(datos_winsor).describe()
})

print("\nComparación de estadísticas descriptivas:")
print(stats_comparacion)

# Efectos en la distribución - visualización
plt.figure(figsize=(15, 10))

# Original vs. Eliminación
plt.subplot(2, 2, 1)
sns.histplot(datos_para_tratamiento, kde=True, bins=50, alpha=0.5, label='Original')
sns.histplot(datos_eliminados, kde=True, bins=50, alpha=0.5, label='Outliers eliminados')
plt.title('Original vs. Eliminación')
plt.legend()
plt.grid(True, alpha=0.3)

# Original vs. Imputación
plt.subplot(2, 2, 2)
sns.histplot(datos_para_tratamiento, kde=True, bins=50, alpha=0.5, label='Original')
sns.histplot(datos_imputados_media, kde=True, bins=50, alpha=0.5, label='Imputación (media)')
sns.histplot(datos_imputados_mediana, kde=True, bins=50, alpha=0.5, label='Imputación (mediana)')
plt.title('Original vs. Imputación')
plt.legend()
plt.grid(True, alpha=0.3)

# Original vs. Winsorización
plt.subplot(2, 2, 3)
sns.histplot(datos_para_tratamiento, kde=True, bins=50, alpha=0.5, label='Original')
sns.histplot(datos_winsor, kde=True, bins=50, alpha=0.5, label='Winsorización')
plt.axvline(lower_bound, color='red', linestyle='--', label=f'Límite inf: {lower_bound:.2f}')
plt.axvline(upper_bound, color='green', linestyle='--', label=f'Límite sup: {upper_bound:.2f}')
plt.title('Original vs. Winsorización')
plt.legend()
plt.grid(True, alpha=0.3)

# Transformaciones
plt.subplot(2, 2, 4)
sns.histplot(datos_log, kde=True, bins=50, alpha=0.5, label='Log')
sns.histplot(datos_sqrt, kde=True, bins=50, alpha=0.5, label='Raíz cuadrada')
sns.histplot(datos_boxcox, kde=True, bins=50, alpha=0.5, label='Box-Cox')
plt.title('Comparación de transformaciones')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Comparación de métodos en un boxplot
plt.figure(figsize=(14, 8))
data_to_plot = [
    datos_para_tratamiento,
    datos_eliminados,
    datos_imputados_media,
    datos_imputados_mediana,
    datos_winsor,
    datos_log,
    datos_sqrt,
    datos_boxcox,
    datos_robustos
]
labels = [
    'Original',
    'Eliminación',
    'Imputación (media)',
    'Imputación (mediana)',
    'Winsorización',
    'Log',
    'Raíz cuadrada',
    'Box-Cox',
    'Escalado robusto'
]

plt.boxplot(data_to_plot, labels=labels, vert=False)
plt.title('Comparación de métodos de tratamiento de outliers')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Efectos en la correlación (usando los datos bivariantes)
# Creamos un nuevo conjunto con outliers en x1 para demostrar
x1_con_outliers = df_biv['x1'].copy()
x2 = df_biv['x2'].copy()

# Aplicar algunos métodos de tratamiento a x1
x1_winsor, _, _ = tratar_outliers_winsor(x1_con_outliers, true_outliers_biv, method='iqr')
x1_imputada = tratar_outliers_imputacion(x1_con_outliers, true_outliers_biv, metodo='mean')
x1_log = tratar_outliers_transformacion(x1_con_outliers, metodo='log')

print("\nEfectos de los tratamientos en la correlación:")
print(f"Correlación original: {np.corrcoef(x1_con_outliers, x2)[0, 1]:.4f}")
print(f"Correlación con winsorización: {np.corrcoef(x1_winsor, x2)[0, 1]:.4f}")
print(f"Correlación con imputación: {np.corrcoef(x1_imputada, x2)[0, 1]:.4f}")
print(f"Correlación con transformación log: {np.corrcoef(x1_log, x2)[0, 1]:.4f}")

# Visualizar el efecto en la correlación
plt.figure(figsize=(15, 5))

# Original
plt.subplot(1, 4, 1)
plt.scatter(x1_con_outliers, x2, alpha=0.5)
plt.title(f'Original\nCorr: {np.corrcoef(x1_con_outliers, x2)[0, 1]:.4f}')
plt.grid(True, alpha=0.3)

# Winsorización
plt.subplot(1, 4, 2)
plt.scatter(x1_winsor, x2, alpha=0.5)
plt.title(f'Winsorización\nCorr: {np.corrcoef(x1_winsor, x2)[0, 1]:.4f}')
plt.grid(True, alpha=0.3)

# Imputación
plt.subplot(1, 4, 3)
plt.scatter(x1_imputada, x2, alpha=0.5)
plt.title(f'Imputación\nCorr: {np.corrcoef(x1_imputada, x2)[0, 1]:.4f}')
plt.grid(True, alpha=0.3)

# Transformación log
plt.subplot(1, 4, 4)
plt.scatter(x1_log, x2, alpha=0.5)
plt.title(f'Log\nCorr: {np.corrcoef(x1_log, x2)[0, 1]:.4f}')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("""
Conclusiones sobre métodos de tratamiento de outliers:
-----------------------------------------------------
1. Eliminación: Simple y efectiva cuando hay pocos outliers, pero reduce el tamaño de la muestra
   y puede introducir sesgos si los datos no son MCAR (Missing Completely At Random).

2. Imputación: Preserva el tamaño de la muestra pero puede distorsionar la distribución y
   subestimar la varianza. La elección del valor de reemplazo (media, mediana, etc.) es crucial.

3. Winsorización (capping): Mantiene los outliers pero limita su influencia, preservando la
   distribución general. Es un buen equilibrio entre eliminación y conservación.

4. Transformaciones: Reducen la asimetría y comprimen el rango de valores extremos. Son útiles
   cuando la distribución subyacente es asimétrica o cuando existen restricciones naturales
   (ej., valores no negativos). La transformación logarítmica es especialmente útil para datos
   con distribución de cola derecha (positivamente sesgada).

5. Métodos robustos: Utilizan estadísticas resistentes a outliers. Son preferibles cuando se
   desea analizar todos los datos pero mitigando la influencia de valores extremos.

La elección depende de:
- El objetivo del análisis
- La proporción de outliers
- La naturaleza de los outliers (errores vs. valores válidos pero extremos)
- El impacto en la distribución y relaciones entre variables
""")

#-----------------------------------------------------------------
# 5. ENFOQUES COMBINADOS Y MEJORES PRÁCTICAS
#-----------------------------------------------------------------

print("\nENFOQUES COMBINADOS Y MEJORES PRÁCTICAS")
print("--------------------------------------")

print("""
Dado que no existe un método universal para la detección y tratamiento de outliers,
a menudo se recomienda un enfoque combinado y sistemático:

1. Investigación preliminar:
   - Análisis exploratorio de datos (EDA)
   - Visualización de distribuciones
   - Comprensión del dominio y el proceso de generación de datos

2. Detección de outliers:
   - Combinación de métodos univariantes para cada variable
   - Métodos multivariantes para capturar relaciones entre variables
   - Validación de los outliers detectados (¿son errores o valores extremos válidos?)

3. Tratamiento de outliers:
   - Tratamiento diferenciado según el tipo de outlier y causa
   - Documentación clara de las decisiones tomadas
   - Validación del impacto del tratamiento en los análisis posteriores

4. Análisis de sensibilidad:
   - Comparación de resultados con y sin tratamiento de outliers
   - Uso de diferentes métodos de tratamiento para evaluar robustez
   - Verificación de que las conclusiones son estables y no dependen del método específico
""")

# Ejemplo de enfoque combinado
def enfoque_combinado_deteccion(datos, umbral_votos=2):
    """
    Aplica múltiples métodos de detección y combina sus resultados.

    Parámetros:
    -----------
    datos : DataFrame, Series o array
        Los datos a analizar
    umbral_votos : int
        Número mínimo de métodos que deben identificar un punto como outlier

    Retorna:
    --------
    outliers_combinados : array
        Índices de las observaciones consideradas outliers por al menos 'umbral_votos' métodos
    votos : array
        Número de votos para cada observación
    """
    # Asegurarse de que datos es un array numpy
    if isinstance(datos, pd.DataFrame):
        X = datos.values
    elif isinstance(datos, pd.Series):
        X = datos.values  # Convertir Series a array NumPy
    else:
        X = datos

    # Si es univariante
    if X.ndim == 1 or (X.ndim == 2 and X.shape[1] == 1):
        X = X.reshape(-1, 1) if X.ndim == 1 else X

        # Métodos univariantes
        outliers_zscore = detectar_outliers_zscore(X.flatten())
        outliers_zscore_mod = detectar_outliers_zscore_modificado(X.flatten())
        outliers_iqr = detectar_outliers_iqr(X.flatten())
        outliers_percentiles = detectar_outliers_percentiles(X.flatten())

        # Contar votos para cada observación
        n_observaciones = X.shape[0]
        votos = np.zeros(n_observaciones)

        for outliers in [outliers_zscore, outliers_zscore_mod, outliers_iqr, outliers_percentiles]:
            votos[outliers] += 1

    else:  # Multivariante
        # Estandarizar
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Métodos multivariantes
        outliers_mahalanobis, _ = detectar_outliers_mahalanobis(X_scaled)
        outliers_pca, _ = detectar_outliers_pca(X_scaled)
        outliers_iforest = detectar_outliers_isolation_forest(X_scaled)
        outliers_lof = detectar_outliers_lof(X_scaled)
        outliers_envelope = detectar_outliers_envelope(X_scaled)

        # Contar votos para cada observación
        n_observaciones = X.shape[0]
        votos = np.zeros(n_observaciones)

        for outliers in [outliers_mahalanobis, outliers_pca, outliers_iforest, outliers_lof, outliers_envelope]:
            votos[outliers] += 1

    # Outliers que superan el umbral de votos
    outliers_combinados = np.where(votos >= umbral_votos)[0]

    return outliers_combinados, votos

# Ejemplo de aplicación del enfoque combinado
outliers_combinados_uni, votos_uni = enfoque_combinado_deteccion(df_uni['valor'], umbral_votos=2)
outliers_combinados_biv, votos_biv = enfoque_combinado_deteccion(df_biv, umbral_votos=2)

print("\nResultados del enfoque combinado:")
print(f"Outliers univariantes detectados por al menos 2 métodos: {len(outliers_combinados_uni)}")
print(f"Outliers bivariantes detectados por al menos 2 métodos: {len(outliers_combinados_biv)}")

# Visualizar los resultados combinados
plt.figure(figsize=(15, 5))

# Univariante
plt.subplot(1, 2, 1)
plt.hist(df_uni['valor'], bins=50, alpha=0.5, label='Datos')
if len(outliers_combinados_uni) > 0:
    plt.hist(df_uni['valor'].iloc[outliers_combinados_uni], bins=50, alpha=0.8, color='red', label='Outliers detectados')
plt.title('Método combinado - Univariante')
plt.legend()
plt.grid(True, alpha=0.3)

# Bivariante
plt.subplot(1, 2, 2)
plt.scatter(df_biv['x1'], df_biv['x2'], alpha=0.5, label='Datos')
if len(outliers_combinados_biv) > 0:
    plt.scatter(df_biv['x1'].iloc[outliers_combinados_biv], df_biv['x2'].iloc[outliers_combinados_biv],
                color='red', alpha=0.8, s=50, label='Outliers detectados')
plt.title('Método combinado - Bivariante')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("""
Mejores prácticas para la detección y tratamiento de outliers:
-------------------------------------------------------------

1. Investigación preliminar:
   - Comprender el dominio y el proceso de generación de datos
   - Determinar si los outliers son errores o valores extremos válidos
   - Analizar el impacto potencial de los outliers en el análisis

2. Detección:
   - Utilizar múltiples métodos y comparar resultados
   - Ajustar los umbrales según el contexto y necesidades específicas
   - Evaluar tanto outliers univariantes como multivariantes
   - Validar visualmente los outliers detectados

3. Tratamiento:
   - Elegir el método de tratamiento según la causa y tipo de outliers
   - Considerar el impacto del tratamiento en distribuciones y relaciones
   - Documentar claramente las decisiones tomadas y sus justificaciones
   - Conservar los datos originales y crear versiones tratadas

4. Validación:
   - Realizar análisis con y sin tratamiento de outliers
   - Evaluar la sensibilidad de los resultados a diferentes métodos
   - Verificar que las conclusiones son robustas

5. Casos especiales:
   - Series temporales: considerar patrones temporales y estacionalidad
   - Alta dimensionalidad: utilizar técnicas específicas como Isolation Forest
   - Outliers categóricos: analizar frecuencias y combinaciones inusuales

Recuerde que el objetivo no es eliminar ciegamente los outliers, sino entenderlos
y mitigar su impacto negativo mientras se preserva la información valiosa que puedan contener.
""")

# Tabla comparativa de métodos
metodos_deteccion = pd.DataFrame({
    'Método': [
        'Z-score', 'Z-score Modificado', 'IQR', 'Análisis de percentiles', 'DBSCAN (univariante)',
        'Distancia de Mahalanobis', 'PCA', 'Isolation Forest', 'LOF', 'Envelope Elíptico', 'DBSCAN (multivariante)'
    ],
    'Tipo': [
        'Univariante', 'Univariante', 'Univariante', 'Univariante', 'Univariante',
        'Multivariante', 'Multivariante', 'Multivariante', 'Multivariante', 'Multivariante', 'Multivariante'
    ],
    'Fortalezas': [
        'Simple, intuitivo',
        'Robusto a múltiples outliers',
        'No asume normalidad, robusto',
        'Flexibilidad para ajustar umbrales',
        'Identifica clusters de densidad',
        'Considera correlaciones entre variables',
        'Reduce dimensionalidad',
        'Eficiente en alta dimensionalidad',
        'Detecta outliers locales',
        'Preciso para distribuciones elípticas',
        'No asume forma específica de clusters'
    ],
    'Debilidades': [
        'Sensible a outliers múltiples',
        'Requiere datos con distribución unimodal',
        'Puede ser demasiado conservador',
        'Requiere ajuste manual de percentiles',
        'Sensible a parámetros',
        'Asume distribución normal multivariante',
        'Puede perder outliers en componentes menores',
        'Resultados pueden variar entre ejecuciones',
        'Sensible al parámetro de vecinos',
        'Estricto con distribución normal',
        'Muy sensible a parámetros'
    ]
})

metodos_tratamiento = pd.DataFrame({
    'Método': [
        'Eliminación', 'Imputación', 'Winsorización', 'Transformación logarítmica',
        'Transformación Box-Cox', 'Escalado robusto'
    ],
    'Aplicación': [
        'Errores claros, pocas observaciones',
        'Mantener tamaño de muestra',
        'Limitar influencia sin eliminar',
        'Distribuciones con asimetría positiva',
        'Normalización de distribuciones',
        'Datos con outliers inevitables'
    ],
    'Ventajas': [
        'Simple, elimina completamente la influencia',
        'Preserva el tamaño de la muestra',
        'Balance entre eliminación y conservación',
        'Reduce asimetría y comprime rangos extremos',
        'Flexibilidad para diferentes distribuciones',
        'Mitiga influencia sin modificar datos'
    ],
    'Desventajas': [
        'Reduce tamaño de muestra, posible sesgo',
        'Distorsiona distribución, reduce varianza',
        'Puede crear acumulación artificial en límites',
        'No adecuada para datos negativos o cero',
        'Interpretabilidad de resultados',
        'Puede no ser suficiente para outliers extremos'
    ]
})

print("\nTabla comparativa de métodos de detección:")
print(metodos_deteccion)

print("\nTabla comparativa de métodos de tratamiento:")
print(metodos_tratamiento)

# Resumen final
print("""
RESUMEN FINAL
------------
La detección y tratamiento de outliers es un proceso que requiere tanto conocimiento técnico
como comprensión del dominio. No existe un enfoque único que funcione en todos los casos,
por lo que es importante:

1. Combinar diferentes métodos de detección
2. Entender la naturaleza y causa de los outliers
3. Seleccionar el tratamiento apropiado según el contexto
4. Validar el impacto del tratamiento en los análisis

Este notebook ha presentado los principales métodos tanto para la detección como para el
tratamiento de outliers, demostrando su aplicación en contextos univariantes y multivariantes.
La elección final del enfoque dependerá de las características específicas de los datos y
los objetivos del análisis.

Recuerde que los outliers no siempre son errores - a veces contienen información valiosa
sobre comportamientos o fenómenos inusuales que merecen ser investigados más a fondo.
""")
```
