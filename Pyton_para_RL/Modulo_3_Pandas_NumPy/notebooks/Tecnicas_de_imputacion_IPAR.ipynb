```python
"""
# Métodos de Imputación de Datos: Teoría y Práctica en Python

Este notebook presenta un estudio completo sobre los métodos de imputación de datos faltantes,
una tarea fundamental en el preprocesamiento de datos para aprendizaje automático y análisis estadístico.

Autor: Dr. Darío Ezequiel Díaz
"""

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer  # Necesario para usar IterativeImputer
from sklearn.impute import SimpleImputer, KNNImputer, IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import missingno as msno
import warnings
warnings.filterwarnings('ignore')

# Configurar el estilo de visualización
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Semilla para reproducibilidad
np.random.seed(42)

#-----------------------------------------------------------------
# 1. INTRODUCCIÓN A LOS DATOS FALTANTES
#-----------------------------------------------------------------

def create_dataset_with_missing_values():
    """
    Crea un conjunto de datos sintético con valores faltantes para demostrar
    las técnicas de imputación.
    """
    # Crear un DataFrame con datos sintéticos
    n_rows = 1000

    # Crear características correlacionadas
    x1 = np.random.normal(0, 1, n_rows)
    x2 = x1 * 0.8 + np.random.normal(0, 0.5, n_rows)  # Correlacionado con x1
    x3 = np.random.normal(5, 2, n_rows)

    # Variable categórica
    categories = ['A', 'B', 'C', 'D']
    cat = np.random.choice(categories, size=n_rows, p=[0.4, 0.3, 0.2, 0.1])

    # Generar valores target basados en las características
    y = 2*x1 - 1.5*x2 + 0.5*x3 + np.random.normal(0, 1, n_rows)

    # Crear el DataFrame
    df = pd.DataFrame({
        'x1': x1,
        'x2': x2,
        'x3': x3,
        'category': cat,
        'target': y
    })

    # Introducir valores faltantes con diferentes patrones

    # 1. Completamente al azar (MCAR - Missing Completely At Random)
    mask_mcar = np.random.random(n_rows) < 0.1
    df.loc[mask_mcar, 'x1'] = np.nan

    # 2. Faltantes según el valor (MAR - Missing At Random)
    # Los valores más altos de x3 tienen más probabilidad de tener x2 faltante
    prob_missing = 1 / (1 + np.exp(-(df['x3'] - df['x3'].mean()) / df['x3'].std()))
    mask_mar = np.random.random(n_rows) < prob_missing * 0.2
    df.loc[mask_mar, 'x2'] = np.nan

    # 3. Faltantes no al azar (MNAR - Missing Not At Random)
    # Los valores más extremos de target tienen más probabilidad de estar faltantes
    mask_mnar = (abs(df['target']) > np.percentile(abs(df['target']), 80))
    df.loc[mask_mnar & (np.random.random(n_rows) < 0.5), 'x3'] = np.nan

    # 4. Valores faltantes en la variable categórica
    mask_cat = np.random.random(n_rows) < 0.15
    df.loc[mask_cat, 'category'] = np.nan

    return df

# Crear el conjunto de datos
data = create_dataset_with_missing_values()

# Visualizar las primeras filas
print("Muestra del conjunto de datos:")
print(data.head())

# Información sobre los datos
print("\nInformación del conjunto de datos:")
print(data.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(data.describe())

# Visualizar los valores faltantes
print("\nCantidad de valores faltantes por columna:")
print(data.isnull().sum())

# Porcentaje de valores faltantes
print("\nPorcentaje de valores faltantes por columna:")
print(data.isnull().mean() * 100)

#-----------------------------------------------------------------
# 2. VISUALIZACIÓN DE VALORES FALTANTES
#-----------------------------------------------------------------

# Visualizar el patrón de valores faltantes
print("\nVisualización de patrones de valores faltantes:")

# Mapa de calor de valores faltantes
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title('Mapa de calor de valores faltantes')
plt.show()

# Matriz de valores faltantes con missingno
msno.matrix(data)
plt.title('Matriz de valores faltantes')
plt.show()

# Dendrograma de valores faltantes
msno.dendrogram(data)
plt.title('Dendograma de valores faltantes')
plt.show()

# Correlación entre valores faltantes
plt.figure(figsize=(8, 6))
sns.heatmap(data.isnull().corr(), annot=True, cmap='coolwarm')
plt.title('Correlación entre valores faltantes')
plt.show()

#-----------------------------------------------------------------
# 3. ELIMINACIÓN DE VALORES FALTANTES
#-----------------------------------------------------------------

# Aunque no es un método de imputación, es importante conocer las técnicas de eliminación

# 3.1 Eliminar filas con valores faltantes (listwise deletion)
data_dropna = data.dropna()
print(f"\nForma original del DataFrame: {data.shape}")
print(f"Forma después de eliminar filas con valores faltantes: {data_dropna.shape}")
print(f"Porcentaje de datos perdidos: {(1 - len(data_dropna) / len(data)) * 100:.2f}%")

# 3.2 Eliminar columnas con muchos valores faltantes (threshold)
threshold = 0.3  # 30% de valores faltantes
data_dropna_cols = data.dropna(axis=1, thresh=int(len(data) * (1 - threshold)))
print(f"\nColumnas originales: {data.columns.tolist()}")
print(f"Columnas después de eliminar aquellas con más de {threshold*100}% de valores faltantes: {data_dropna_cols.columns.tolist()}")

# 3.3 Eliminar filas con valores faltantes en columnas específicas
data_dropna_subset = data.dropna(subset=['target'])
print(f"\nForma después de eliminar filas con valores faltantes en 'target': {data_dropna_subset.shape}")

# Ventajas y desventajas de la eliminación
print("""
Ventajas de la eliminación:
- Simplicidad: fácil de implementar
- No introduce sesgos si los datos faltan completamente al azar (MCAR)

Desventajas de la eliminación:
- Pérdida de información y poder estadístico
- Potencial sesgo si los datos no faltan completamente al azar
- Ineficiente si hay muchos valores faltantes
""")

#-----------------------------------------------------------------
# 4. MÉTODOS DE IMPUTACIÓN UNIVARIADA
#-----------------------------------------------------------------

# Crear una copia del conjunto de datos original para cada método
data_original = data.copy()

# 4.1 Imputación con estadísticas descriptivas (media, mediana, moda)

# Imputación con media
data_mean = data_original.copy()
data_mean['x1'].fillna(data_mean['x1'].mean(), inplace=True)
data_mean['x2'].fillna(data_mean['x2'].mean(), inplace=True)
data_mean['x3'].fillna(data_mean['x3'].mean(), inplace=True)

# Imputación con mediana
data_median = data_original.copy()
data_median['x1'].fillna(data_median['x1'].median(), inplace=True)
data_median['x2'].fillna(data_median['x2'].median(), inplace=True)
data_median['x3'].fillna(data_median['x3'].median(), inplace=True)

# Imputación con moda
data_mode = data_original.copy()
data_mode['category'].fillna(data_mode['category'].mode()[0], inplace=True)

# Verificar las imputaciones
print("\nEstadísticas después de la imputación con media:")
print(data_mean.describe()[['x1', 'x2', 'x3']])

print("\nEstadísticas después de la imputación con mediana:")
print(data_median.describe()[['x1', 'x2', 'x3']])

print("\nDistribución de categorías después de la imputación con moda:")
print(data_mode['category'].value_counts(normalize=True))

# Visualizar la distribución antes y después de la imputación
plt.figure(figsize=(15, 10))

# Visualizar x1
plt.subplot(3, 2, 1)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_mean['x1'], label='Imputado con Media')
plt.title('Distribución de x1')
plt.legend()

plt.subplot(3, 2, 2)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_median['x1'], label='Imputado con Mediana')
plt.title('Distribución de x1')
plt.legend()

# Visualizar x2
plt.subplot(3, 2, 3)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_mean['x2'], label='Imputado con Media')
plt.title('Distribución de x2')
plt.legend()

plt.subplot(3, 2, 4)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_median['x2'], label='Imputado con Mediana')
plt.title('Distribución de x2')
plt.legend()

# Visualizar x3
plt.subplot(3, 2, 5)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_mean['x3'], label='Imputado con Media')
plt.title('Distribución de x3')
plt.legend()

plt.subplot(3, 2, 6)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_median['x3'], label='Imputado con Mediana')
plt.title('Distribución de x3')
plt.legend()

plt.tight_layout()
plt.show()

# 4.2 Imputación con valores constantes

# Imputación con cero
data_zero = data_original.copy()
data_zero[['x1', 'x2', 'x3']] = data_zero[['x1', 'x2', 'x3']].fillna(0)

# Imputación con un valor constante predeterminado
data_const = data_original.copy()
data_const['x1'].fillna(-999, inplace=True)
data_const['x2'].fillna(-999, inplace=True)
data_const['x3'].fillna(-999, inplace=True)
data_const['category'].fillna('MISSING', inplace=True)

print("\nEstadísticas después de la imputación con cero:")
print(data_zero.describe()[['x1', 'x2', 'x3']])

print("\nEstadísticas después de la imputación con constante (-999):")
print(data_const.describe()[['x1', 'x2', 'x3']])

print("\nDistribución de categorías después de la imputación con 'MISSING':")
print(data_const['category'].value_counts())

# 4.3 Imputación con interpolación (para series temporales o datos ordenados)

# Para demostrar esto, vamos a crear una serie temporal simple
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
ts_data = pd.Series(np.sin(np.linspace(0, 15, 100)) + np.random.normal(0, 0.1, 100), index=dates)

# Introducir valores faltantes
missing_indices = np.random.choice(range(1, 99), 20, replace=False)
ts_data[missing_indices] = np.nan

# Diferentes métodos de interpolación
ts_linear = ts_data.interpolate(method='linear')
ts_spline = ts_data.interpolate(method='spline', order=3)
ts_polynomial = ts_data.interpolate(method='polynomial', order=2)

# Visualizar los resultados
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
ts_data.plot(marker='o', linestyle='None', markersize=4, label='Original con NaNs')
plt.title('Serie temporal con valores faltantes')
plt.legend()

plt.subplot(4, 1, 2)
ts_data.plot(marker='o', linestyle='None', markersize=4, color='red', alpha=0.5)
ts_linear.plot(marker='', linestyle='-', label='Interpolación Lineal')
plt.title('Imputación con Interpolación Lineal')
plt.legend()

plt.subplot(4, 1, 3)
ts_data.plot(marker='o', linestyle='None', markersize=4, color='red', alpha=0.5)
ts_spline.plot(marker='', linestyle='-', label='Interpolación Spline')
plt.title('Imputación con Interpolación Spline (cúbica)')
plt.legend()

plt.subplot(4, 1, 4)
ts_data.plot(marker='o', linestyle='None', markersize=4, color='red', alpha=0.5)
ts_polynomial.plot(marker='', linestyle='-', label='Interpolación Polinómica')
plt.title('Imputación con Interpolación Polinómica (order=2)')
plt.legend()

plt.tight_layout()
plt.show()

# 4.4 Imputación mediante propagación (forward fill y backward fill)

# Forward Fill (ffill) - propagar el último valor válido
data_ffill = data_original.copy()
data_ffill.ffill(inplace=True)

# Backward Fill (bfill) - propagar el siguiente valor válido
data_bfill = data_original.copy()
data_bfill.bfill(inplace=True)

# Como estos métodos son más útiles para datos secuenciales, los demostramos con la serie temporal
plt.figure(figsize=(12, 9))

plt.subplot(3, 1, 1)
ts_data.plot(marker='o', linestyle='None', markersize=4, label='Original con NaNs')
plt.title('Serie temporal con valores faltantes')
plt.legend()

plt.subplot(3, 1, 2)
ts_data.plot(marker='o', linestyle='None', markersize=4, color='red', alpha=0.5)
ts_data.ffill().plot(marker='', linestyle='-', label='Forward Fill')
plt.title('Imputación con Forward Fill (propagar el último valor válido)')
plt.legend()

plt.subplot(3, 1, 3)
ts_data.plot(marker='o', linestyle='None', markersize=4, color='red', alpha=0.5)
ts_data.bfill().plot(marker='', linestyle='-', label='Backward Fill')
plt.title('Imputación con Backward Fill (propagar el siguiente valor válido)')
plt.legend()

plt.tight_layout()
plt.show()

# Ventajas y desventajas de los métodos univariados
print("""
Ventajas de los métodos univariados:
- Simples y rápidos de implementar
- No requieren otras variables para la imputación
- Buenos para conjuntos de datos pequeños con pocos valores faltantes

Desventajas de los métodos univariados:
- No consideran las relaciones entre variables
- Pueden distorsionar la distribución y correlaciones de los datos
- La imputación con media/mediana reduce la varianza
- Los valores constantes pueden afectar negativamente a algunos algoritmos
""")

#-----------------------------------------------------------------
# 5. MÉTODOS DE IMPUTACIÓN MULTIVARIADA
#-----------------------------------------------------------------

# 5.1 Imputación con K-Nearest Neighbors (KNN)

# Separar características numéricas y categóricas
numeric_cols = ['x1', 'x2', 'x3', 'target']
data_numeric = data_original[numeric_cols].copy()

# Escalar los datos para KNN
scaler = StandardScaler()
data_numeric_scaled = pd.DataFrame(
    scaler.fit_transform(data_numeric.dropna()),
    columns=numeric_cols,
    index=data_numeric.dropna().index
)

# Aplicar KNN Imputer con diferentes valores de k
knn_imputer_3 = KNNImputer(n_neighbors=3)
knn_imputer_5 = KNNImputer(n_neighbors=5)
knn_imputer_10 = KNNImputer(n_neighbors=10)

# Imputar los datos escalados
data_knn_3_scaled = pd.DataFrame(
    knn_imputer_3.fit_transform(data_numeric),
    columns=numeric_cols,
    index=data_numeric.index
)
data_knn_5_scaled = pd.DataFrame(
    knn_imputer_5.fit_transform(data_numeric),
    columns=numeric_cols,
    index=data_numeric.index
)
data_knn_10_scaled = pd.DataFrame(
    knn_imputer_10.fit_transform(data_numeric),
    columns=numeric_cols,
    index=data_numeric.index
)

# Desescalar los datos
data_knn_3 = pd.DataFrame(
    scaler.inverse_transform(data_knn_3_scaled),
    columns=numeric_cols,
    index=data_numeric.index
)
data_knn_5 = pd.DataFrame(
    scaler.inverse_transform(data_knn_5_scaled),
    columns=numeric_cols,
    index=data_numeric.index
)
data_knn_10 = pd.DataFrame(
    scaler.inverse_transform(data_knn_10_scaled),
    columns=numeric_cols,
    index=data_numeric.index
)

# Visualizar los resultados para una variable
plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_knn_3['x2'], label='KNN (k=3)')
plt.title('Imputación KNN (k=3) para x2')
plt.legend()

plt.subplot(1, 3, 2)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_knn_5['x2'], label='KNN (k=5)')
plt.title('Imputación KNN (k=5) para x2')
plt.legend()

plt.subplot(1, 3, 3)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_knn_10['x2'], label='KNN (k=10)')
plt.title('Imputación KNN (k=10) para x2')
plt.legend()

plt.tight_layout()
plt.show()

# 5.2 Imputación mediante modelos de regresión

# Preparar los datos para imputación por regresión (solo para x1, x2, x3)
data_for_imputation = data_original[['x1', 'x2', 'x3']].copy()

# Imputación usando IterativeImputer con diferentes estimadores
# Linear Regression
imputer_lr = IterativeImputer(
    estimator=LinearRegression(),
    max_iter=10,
    random_state=42
)
data_imputed_lr = pd.DataFrame(
    imputer_lr.fit_transform(data_for_imputation),
    columns=data_for_imputation.columns,
    index=data_for_imputation.index
)

# BayesianRidge (método por defecto)
imputer_br = IterativeImputer(
    estimator=BayesianRidge(),
    max_iter=10,
    random_state=42
)
data_imputed_br = pd.DataFrame(
    imputer_br.fit_transform(data_for_imputation),
    columns=data_for_imputation.columns,
    index=data_for_imputation.index
)

# RandomForest
imputer_rf = IterativeImputer(
    estimator=RandomForestRegressor(n_estimators=100, random_state=42),
    max_iter=10,
    random_state=42
)
data_imputed_rf = pd.DataFrame(
    imputer_rf.fit_transform(data_for_imputation),
    columns=data_for_imputation.columns,
    index=data_for_imputation.index
)

# Visualizar los resultados
plt.figure(figsize=(15, 12))

# Para x1
plt.subplot(3, 3, 1)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_lr['x1'], label='Regresión Lineal')
plt.title('Imputación con Regresión Lineal (x1)')
plt.legend()

plt.subplot(3, 3, 2)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_br['x1'], label='Bayesian Ridge')
plt.title('Imputación con Bayesian Ridge (x1)')
plt.legend()

plt.subplot(3, 3, 3)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_rf['x1'], label='Random Forest')
plt.title('Imputación con Random Forest (x1)')
plt.legend()

# Para x2
plt.subplot(3, 3, 4)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_lr['x2'], label='Regresión Lineal')
plt.title('Imputación con Regresión Lineal (x2)')
plt.legend()

plt.subplot(3, 3, 5)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_br['x2'], label='Bayesian Ridge')
plt.title('Imputación con Bayesian Ridge (x2)')
plt.legend()

plt.subplot(3, 3, 6)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_rf['x2'], label='Random Forest')
plt.title('Imputación con Random Forest (x2)')
plt.legend()

# Para x3
plt.subplot(3, 3, 7)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_lr['x3'], label='Regresión Lineal')
plt.title('Imputación con Regresión Lineal (x3)')
plt.legend()

plt.subplot(3, 3, 8)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_br['x3'], label='Bayesian Ridge')
plt.title('Imputación con Bayesian Ridge (x3)')
plt.legend()

plt.subplot(3, 3, 9)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_imputed_rf['x3'], label='Random Forest')
plt.title('Imputación con Random Forest (x3)')
plt.legend()

plt.tight_layout()
plt.show()

# 5.3 Imputación multivariante con PCA

# Función para imputación con PCA
def pca_imputation(data, n_components=2, max_iter=10, tol=1e-6):
    """
    Realiza imputación de valores faltantes utilizando PCA iterativo.

    Parámetros:
    -----------
    data : DataFrame
        Los datos con valores faltantes.
    n_components : int
        Número de componentes principales a utilizar.
    max_iter : int
        Número máximo de iteraciones.
    tol : float
        Tolerancia para la convergencia.

    Retorna:
    --------
    DataFrame
        Los datos con valores imputados.
    """
    # Crear una copia de los datos
    imputed_data = data.copy()

    # Imputar primero con la media para tener un punto de partida
    for col in imputed_data.columns:
        imputed_data[col].fillna(imputed_data[col].mean(), inplace=True)

    # Matriz de máscaras para valores faltantes
    mask = data.isna()

    # Estandarizar los datos
    scaler = StandardScaler()
    imputed_scaled = scaler.fit_transform(imputed_data)

    # Iteración para PCA
    for i in range(max_iter):
        # Guardar la versión anterior para verificar convergencia
        prev_imputed = imputed_scaled.copy()

        # Ajustar PCA
        pca = PCA(n_components=n_components)
        pca.fit(imputed_scaled)

        # Transformar y luego reconstruir
        transformed = pca.transform(imputed_scaled)
        reconstructed = pca.inverse_transform(transformed)

        # Actualizar solo los valores faltantes
        for j, col in enumerate(data.columns):
            missing_idx = mask[col].values
            imputed_scaled[missing_idx, j] = reconstructed[missing_idx, j]

        # Verificar convergencia
        diff = np.mean(np.abs(imputed_scaled - prev_imputed))
        if diff < tol:
            break

    # Desescalar y convertir de nuevo a DataFrame
    imputed_final = scaler.inverse_transform(imputed_scaled)
    imputed_df = pd.DataFrame(imputed_final, columns=data.columns, index=data.index)

    return imputed_df

# Aplicar imputación PCA
data_pca_2 = pca_imputation(data_for_imputation, n_components=2)
data_pca_3 = pca_imputation(data_for_imputation, n_components=3)

# Visualizar los resultados
plt.figure(figsize=(15, 6))

# Para x1
plt.subplot(1, 3, 1)
sns.kdeplot(data=data_original['x1'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_pca_2['x1'], label='PCA (n=2)')
sns.kdeplot(data=data_pca_3['x1'], label='PCA (n=3)')
plt.title('Imputación con PCA (x1)')
plt.legend()

# Para x2
plt.subplot(1, 3, 2)
sns.kdeplot(data=data_original['x2'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_pca_2['x2'], label='PCA (n=2)')
sns.kdeplot(data=data_pca_3['x2'], label='PCA (n=3)')
plt.title('Imputación con PCA (x2)')
plt.legend()

# Para x3
plt.subplot(1, 3, 3)
sns.kdeplot(data=data_original['x3'].dropna(), label='Original (sin NaNs)')
sns.kdeplot(data=data_pca_2['x3'], label='PCA (n=2)')
sns.kdeplot(data=data_pca_3['x3'], label='PCA (n=3)')
plt.title('Imputación con PCA (x3)')
plt.legend()

plt.tight_layout()
plt.show()

# Ventajas y desventajas de los métodos multivariados
print("""
Ventajas de los métodos multivariados:
- Consideran las relaciones entre variables
- Pueden preservar mejor la estructura de covarianza de los datos
- Generalmente producen imputaciones más precisas
- Adecuados para patrones de valores faltantes MAR o MNAR

Desventajas de los métodos multivariados:
- Mayor complejidad computacional
- Requieren ajuste de hiperparámetros
- Pueden sobreajustar en conjuntos de datos pequeños
- La imputación puede fallar si hay muchos valores faltantes
""")

#-----------------------------------------------------------------
# 6. EVALUACIÓN DE MÉTODOS DE IMPUTACIÓN
#-----------------------------------------------------------------

# 6.1 Introducir valores faltantes artificiales en un conjunto completo

def evaluate_imputation(data_complete, missing_rate=0.1, n_splits=5):
    """
    Evalúa diferentes métodos de imputación utilizando datos completos
    e introduciendo valores faltantes artificiales.

    Parámetros:
    -----------
    data_complete : DataFrame
        Los datos completos sin valores faltantes.
    missing_rate : float
        Tasa de valores faltantes a introducir.
    n_splits : int
        Número de repeticiones/divisiones para la evaluación.

    Retorna:
    --------
    dict
        Un diccionario con los errores medios para cada método.
    """
    # Asegurarnos de trabajar solo con columnas numéricas
    numeric_data = data_complete.select_dtypes(include=['number'])

    if numeric_data.empty:
        print("No hay columnas numéricas para evaluar")
        return {}

    results = {
        'Mean': [],
        'Median': [],
        'KNN (k=5)': [],
        'Linear Regression': [],
        'Random Forest': [],
        'PCA (n=2)': []
    }

    for i in range(n_splits):
        # Crear una máscara de valores faltantes
        mask = np.random.random(numeric_data.shape) < missing_rate

        # Crear datos con valores faltantes
        data_missing = numeric_data.copy()
        data_missing.values[mask] = np.nan

        # Guardar los valores originales para comparación
        true_values = numeric_data.values[mask]

        # Imputación con media
        data_mean = data_missing.copy()
        for col in data_mean.columns:
            data_mean[col].fillna(data_mean[col].mean(), inplace=True)
        mean_values = data_mean.values[mask]
        results['Mean'].append(mean_squared_error(true_values, mean_values))

        # Imputación con mediana
        data_median = data_missing.copy()
        for col in data_median.columns:
            data_median[col].fillna(data_median[col].median(), inplace=True)
        median_values = data_median.values[mask]
        results['Median'].append(mean_squared_error(true_values, median_values))

        # Imputación con KNN
        knn_imputer = KNNImputer(n_neighbors=5)
        data_knn = pd.DataFrame(
            knn_imputer.fit_transform(data_missing),
            columns=data_missing.columns,
            index=data_missing.index
        )
        knn_values = data_knn.values[mask]
        results['KNN (k=5)'].append(mean_squared_error(true_values, knn_values))

        # Imputación con regresión lineal
        imputer_lr = IterativeImputer(
            estimator=LinearRegression(),
            max_iter=10,
            random_state=42
        )
        data_lr = pd.DataFrame(
            imputer_lr.fit_transform(data_missing),
            columns=data_missing.columns,
            index=data_missing.index
        )
        lr_values = data_lr.values[mask]
        results['Linear Regression'].append(mean_squared_error(true_values, lr_values))

        # Imputación con Random Forest
        imputer_rf = IterativeImputer(
            estimator=RandomForestRegressor(n_estimators=100, random_state=42),
            max_iter=10,
            random_state=42
        )
        data_rf = pd.DataFrame(
            imputer_rf.fit_transform(data_missing),
            columns=data_missing.columns,
            index=data_missing.index
        )
        rf_values = data_rf.values[mask]
        results['Random Forest'].append(mean_squared_error(true_values, rf_values))

        # Imputación con PCA
        data_pca = pca_imputation(data_missing, n_components=2)
        pca_values = data_pca.values[mask]
        results['PCA (n=2)'].append(mean_squared_error(true_values, pca_values))

    # Calcular el promedio de los errores
    for method in results:
        results[method] = np.mean(results[method])

    return results

# Para esta evaluación, necesitamos un conjunto de datos completo
# Usaremos solo las filas sin valores faltantes
data_complete = data_original.dropna()

# Si hay muy pocas filas completas, podemos considerar solo un subconjunto de variables
if len(data_complete) < 100:
    # Usar solo variables numéricas con menos valores faltantes
    cols_to_use = ['x1', 'x2', 'x3']
    mask = data_original[cols_to_use].isna().sum(axis=1) == 0
    data_complete = data_original.loc[mask, cols_to_use]

print(f"Número de filas completas para evaluación: {len(data_complete)}")

# Realizar la evaluación
if len(data_complete) >= 50:  # Asegurarse de que hay suficientes datos
    errors = evaluate_imputation(data_complete, missing_rate=0.2, n_splits=5)

    # Visualizar los resultados
    plt.figure(figsize=(10, 6))
    methods = list(errors.keys())
    error_values = list(errors.values())

    # Ordenar por error (menor a mayor)
    sorted_indices = np.argsort(error_values)
    methods = [methods[i] for i in sorted_indices]
    error_values = [error_values[i] for i in sorted_indices]

    plt.barh(methods, error_values, color='skyblue')
    plt.xlabel('Error Cuadrático Medio')
    plt.title('Comparación de Métodos de Imputación')
    plt.tight_layout()
    plt.show()

    print("\nResultados de la evaluación (MSE):")
    for method, error in zip(methods, error_values):
        print(f"{method}: {error:.6f}")
else:
    print("No hay suficientes datos completos para realizar una evaluación confiable.")

# 6.2 Impacto en la predicción: evaluación basada en tareas

def evaluate_imputation_predictive(data, target_col='target', test_size=0.2):
    """
    Evalúa el impacto de diferentes métodos de imputación en una tarea de predicción.

    Parámetros:
    -----------
    data : DataFrame
        Los datos con valores faltantes.
    target_col : str
        El nombre de la columna objetivo.
    test_size : float
        Proporción de datos a utilizar como conjunto de prueba.

    Retorna:
    --------
    dict
        Un diccionario con los errores de predicción para cada método.
    """
    # Separar características y variable objetivo
    X = data.drop(columns=[target_col])
    y = data[target_col]

    # Identificar columnas numéricas y categóricas
    numeric_features = X.select_dtypes(include=['number']).columns.tolist()
    categorical_features = X.select_dtypes(exclude=['number']).columns.tolist()

    # División en conjuntos de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42)

    # Resultados para cada método
    results = {}

    # Modelo de predicción (Random Forest)
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

    # 1. Método Dropna (eliminar filas con valores faltantes)
    try:
        # Eliminar filas con valores faltantes
        X_train_dropna = X_train.dropna()
        y_train_dropna = y_train.loc[X_train_dropna.index]
        X_test_dropna = X_test.dropna()
        y_test_dropna = y_test.loc[X_test_dropna.index]

        # Si hay suficientes datos después de eliminar filas
        if len(X_train_dropna) > 50 and len(X_test_dropna) > 20:
            # Procesamiento de características
            numeric_transformer = Pipeline(steps=[
                ('scaler', StandardScaler())
            ])
            categorical_transformer = Pipeline(steps=[
                ('onehot', OneHotEncoder(handle_unknown='ignore'))
            ])
            preprocessor = ColumnTransformer(transformers=[
                ('num', numeric_transformer, numeric_features),
                ('cat', categorical_transformer, categorical_features) if categorical_features else None
            ], remainder='passthrough')

            # Pipeline completo
            pipeline = Pipeline(steps=[
                ('preprocessor', preprocessor),
                ('model', rf_model)
            ])

            # Entrenar y evaluar
            pipeline.fit(X_train_dropna, y_train_dropna)
            y_pred = pipeline.predict(X_test_dropna)
            mse = mean_squared_error(y_test_dropna, y_pred)
            mae = mean_absolute_error(y_test_dropna, y_pred)
            results['Dropna'] = {'MSE': mse, 'MAE': mae}
        else:
            results['Dropna'] = "No hay suficientes datos después de eliminar filas"
    except Exception as e:
        results['Dropna'] = f"Error: {str(e)}"

    # 2. Método de Imputación con Media/Moda
    try:
        # Preprocesador con imputación media/moda
        numeric_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ])
        categorical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])

        transformers = []
        if numeric_features:
            transformers.append(('num', numeric_transformer, numeric_features))
        if categorical_features:
            transformers.append(('cat', categorical_transformer, categorical_features))

        preprocessor = ColumnTransformer(transformers=transformers)

        # Pipeline completo
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('model', rf_model)
        ])

        # Entrenar y evaluar
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        results['Mean/Mode'] = {'MSE': mse, 'MAE': mae}
    except Exception as e:
        results['Mean/Mode'] = f"Error: {str(e)}"

    # 3. Método de Imputación KNN
    try:
        # Preprocesador con imputación KNN
        numeric_transformer = Pipeline(steps=[
            ('imputer', KNNImputer(n_neighbors=5)),
            ('scaler', StandardScaler())
        ])
        categorical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])

        transformers = []
        if numeric_features:
            transformers.append(('num', numeric_transformer, numeric_features))
        if categorical_features:
            transformers.append(('cat', categorical_transformer, categorical_features))

        preprocessor = ColumnTransformer(transformers=transformers)

        # Pipeline completo
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('model', rf_model)
        ])

        # Entrenar y evaluar
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        results['KNN'] = {'MSE': mse, 'MAE': mae}
    except Exception as e:
        results['KNN'] = f"Error: {str(e)}"

    # 4. Método de Imputación Iterativa (Regresión)
    try:
        if numeric_features:  # Solo si hay características numéricas
            # Preprocesador con imputación iterativa
            numeric_transformer = Pipeline(steps=[
                ('imputer', IterativeImputer(estimator=LinearRegression(), max_iter=10, random_state=42)),
                ('scaler', StandardScaler())
            ])
            categorical_transformer = Pipeline(steps=[
                ('imputer', SimpleImputer(strategy='most_frequent')),
                ('onehot', OneHotEncoder(handle_unknown='ignore'))
            ])

            transformers = []
            if numeric_features:
                transformers.append(('num', numeric_transformer, numeric_features))
            if categorical_features:
                transformers.append(('cat', categorical_transformer, categorical_features))

            preprocessor = ColumnTransformer(transformers=transformers)

            # Pipeline completo
            pipeline = Pipeline(steps=[
                ('preprocessor', preprocessor),
                ('model', rf_model)
            ])

            # Entrenar y evaluar
            pipeline.fit(X_train, y_train)
            y_pred = pipeline.predict(X_test)
            mse = mean_squared_error(y_test, y_pred)
            mae = mean_absolute_error(y_test, y_pred)
            results['Iterative (Regression)'] = {'MSE': mse, 'MAE': mae}
        else:
            results['Iterative (Regression)'] = "No hay características numéricas para aplicar este método"
    except Exception as e:
        results['Iterative (Regression)'] = f"Error: {str(e)}"

    # 5. Método de Imputación Iterativa (Random Forest)
    try:
        if numeric_features:  # Solo si hay características numéricas
            # Preprocesador con imputación iterativa usando Random Forest
            numeric_transformer = Pipeline(steps=[
                ('imputer', IterativeImputer(
                    estimator=RandomForestRegressor(n_estimators=100, random_state=42),
                    max_iter=10, random_state=42)),
                ('scaler', StandardScaler())
            ])
            categorical_transformer = Pipeline(steps=[
                ('imputer', SimpleImputer(strategy='most_frequent')),
                ('onehot', OneHotEncoder(handle_unknown='ignore'))
            ])

            transformers = []
            if numeric_features:
                transformers.append(('num', numeric_transformer, numeric_features))
            if categorical_features:
                transformers.append(('cat', categorical_transformer, categorical_features))

            preprocessor = ColumnTransformer(transformers=transformers)

            # Pipeline completo
            pipeline = Pipeline(steps=[
                ('preprocessor', preprocessor),
                ('model', rf_model)
            ])

            # Entrenar y evaluar
            pipeline.fit(X_train, y_train)
            y_pred = pipeline.predict(X_test)
            mse = mean_squared_error(y_test, y_pred)
            mae = mean_absolute_error(y_test, y_pred)
            results['Iterative (Random Forest)'] = {'MSE': mse, 'MAE': mae}
        else:
            results['Iterative (Random Forest)'] = "No hay características numéricas para aplicar este método"
    except Exception as e:
        results['Iterative (Random Forest)'] = f"Error: {str(e)}"

    return results

# Evaluación del impacto en la predicción
prediction_results = evaluate_imputation_predictive(data_original, target_col='target')

# Visualizar los resultados
plt.figure(figsize=(12, 6))

# Recopilar métodos y valores para MSE
methods_mse = []
mse_values = []
for method, result in prediction_results.items():
    if isinstance(result, dict) and 'MSE' in result:
        methods_mse.append(method)
        mse_values.append(result['MSE'])

# Ordenar por MSE
if methods_mse:
    # MSE
    plt.subplot(1, 2, 1)
    sorted_indices = np.argsort(mse_values)
    methods_sorted = [methods_mse[i] for i in sorted_indices]
    mse_sorted = [mse_values[i] for i in sorted_indices]

    plt.barh(methods_sorted, mse_sorted, color='skyblue')
    plt.xlabel('Error Cuadrático Medio (MSE)')
    plt.title('Impacto en la Predicción (MSE)')

    # MAE
    plt.subplot(1, 2, 2)
    methods_mae = []
    mae_values = []
    for method, result in prediction_results.items():
        if isinstance(result, dict) and 'MAE' in result:
            methods_mae.append(method)
            mae_values.append(result['MAE'])

    sorted_indices = np.argsort(mae_values)
    methods_sorted = [methods_mae[i] for i in sorted_indices]
    mae_sorted = [mae_values[i] for i in sorted_indices]

    plt.barh(methods_sorted, mae_sorted, color='lightgreen')
    plt.xlabel('Error Absoluto Medio (MAE)')
    plt.title('Impacto en la Predicción (MAE)')

    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes resultados para visualizar.")

# Imprimir resultados
print("\nResultados de la evaluación predictiva:")
for method, result in prediction_results.items():
    if isinstance(result, dict) and 'MSE' in result:
        print(f"{method}: MSE = {result['MSE']:.4f}, MAE = {result['MAE']:.4f}")
    else:
        print(f"{method}: {result}")

#-----------------------------------------------------------------
# 7. GUÍA DE MEJORES PRÁCTICAS PARA LA IMPUTACIÓN
#-----------------------------------------------------------------

print("""
\n** GUÍA DE MEJORES PRÁCTICAS PARA LA IMPUTACIÓN DE DATOS **

1. Análisis Exploratorio de Valores Faltantes:
   - Comprenda el patrón de valores faltantes (MCAR, MAR, MNAR)
   - Visualice los valores faltantes para identificar patrones
   - Evalúe la proporción de valores faltantes por variable y observación

2. Selección del Método de Imputación:
   - Para pocos valores faltantes (<5%) y MCAR:
     * Eliminar filas/columnas puede ser apropiado
     * Imputación simple con media/mediana/moda
   - Para valores faltantes moderados (5-15%) o MAR:
     * Imputación basada en KNN
     * Imputación mediante regresión o métodos iterativos
   - Para valores faltantes elevados (>15%) o MNAR:
     * Modelos avanzados como MICE o enfoques de máxima verosimilitud
     * Considerar la imputación múltiple

3. Consideraciones por Tipo de Variable:
   - Variables numéricas:
     * Cuantitativas continuas: media, interpolación, métodos basados en regresión
     * Cuantitativas discretas: mediana, moda, round después de la imputación
   - Variables categóricas:
     * Nominales: moda, modelos de clasificación específicos
     * Ordinales: mediana, preservación del orden

4. Datos Temporales:
   - Series temporales: interpolación, métodos específicos para series temporales
   - Datos secuenciales: forward fill, backward fill, métodos de suavizado

5. Validación de la Imputación:
   - Evaluación en datos sintéticos con valores faltantes artificiales
   - Comparación del impacto en tareas predictivas
   - Análisis de la distribución antes y después de la imputación

6. Implementación en el Flujo de Trabajo:
   - Incluir la imputación dentro del pipeline de procesamiento
   - Aplicar la misma estrategia de imputación en entrenamiento y producción
   - Documentar las decisiones de imputación

7. Consideraciones Avanzadas:
   - Tratar los valores faltantes como una característica informativa
   - Crear indicadores de imputación (dummies) para valores faltantes
   - Considerar la incertidumbre de la imputación en análisis posteriores

8. Herramientas Recomendadas:
   - Pandas: fillna(), interpolate() para métodos básicos
   - Scikit-learn: SimpleImputer, KNNImputer, IterativeImputer
   - Bibliotecas especializadas: missingno, fancyimpute, impyute
""")

#-----------------------------------------------------------------
# 8. EJEMPLO DE APLICACIÓN COMPLETO: FLUJO DE TRABAJO DE IMPUTACIÓN
#-----------------------------------------------------------------

# Un ejemplo completo de flujo de trabajo de imputación en un pipeline de scikit-learn

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

# Crear un conjunto de datos más realista
np.random.seed(42)
n_samples = 1000

# Crear variables categóricas y numéricas con correlaciones
# Categóricas
cat1 = np.random.choice(['A', 'B', 'C'], size=n_samples, p=[0.5, 0.3, 0.2])
cat2 = np.random.choice(['X', 'Y', 'Z'], size=n_samples, p=[0.2, 0.5, 0.3])

# Numéricas
num1 = np.random.normal(0, 1, n_samples)
num2 = 0.7 * num1 + np.random.normal(0, 0.5, n_samples)
num3 = 5 + 0.3 * num1 - 0.2 * num2 + np.random.normal(0, 1, n_samples)
num4 = np.random.normal(10, 2, n_samples)

# Variable objetivo
target = 2*num1 - 1.5*num2 + num3 + 0.5*num4 + np.random.normal(0, 2, n_samples)
target += np.where(cat1 == 'A', 2, np.where(cat1 == 'B', -1, 0))
target += np.where(cat2 == 'X', 1, np.where(cat2 == 'Y', -1, 0))

# Crear DataFrame
df = pd.DataFrame({
    'cat1': cat1,
    'cat2': cat2,
    'num1': num1,
    'num2': num2,
    'num3': num3,
    'num4': num4,
    'target': target
})

# Introducir valores faltantes
# MCAR para cat1 y num1
mask_mcar1 = np.random.random(n_samples) < 0.1
mask_mcar2 = np.random.random(n_samples) < 0.15
df.loc[mask_mcar1, 'cat1'] = np.nan
df.loc[mask_mcar2, 'num1'] = np.nan

# MAR para cat2 (más valores faltantes si num4 es alto)
prob_missing = 1 / (1 + np.exp(-(df['num4'] - df['num4'].mean()) / df['num4'].std()))
mask_mar = np.random.random(n_samples) < prob_missing * 0.2
df.loc[mask_mar, 'cat2'] = np.nan

# MNAR para num3 (más valores faltantes si target es alto)
mask_mnar = (abs(df['target']) > np.percentile(abs(df['target']), 85))
df.loc[mask_mnar & (np.random.random(n_samples) < 0.4), 'num3'] = np.nan

# Visualizar los datos y valores faltantes
print("\nResumen del conjunto de datos de aplicación:")
print(df.info())
print("\nPatrón de valores faltantes:")
print(df.isnull().sum())
print("\nPorcentaje de valores faltantes:")
print(df.isnull().mean() * 100)

# Visualizar matriz de valores faltantes
plt.figure(figsize=(10, 6))
msno.matrix(df)
plt.title('Matriz de Valores Faltantes - Conjunto de Aplicación')
plt.show()

# Definir preprocesadores para variables numéricas y categóricas
numeric_features = ['num1', 'num2', 'num3', 'num4']
categorical_features = ['cat1', 'cat2']

# Pipeline para variables numéricas (con diferentes métodos de imputación)
numeric_transformer_mean = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

numeric_transformer_knn = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler())
])

numeric_transformer_iterative = Pipeline(steps=[
    ('imputer', IterativeImputer(
        estimator=RandomForestRegressor(n_estimators=100, random_state=42),
        random_state=42)),
    ('scaler', StandardScaler())
])

# Pipeline para variables categóricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Crear diferentes preprocesadores completos
preprocessor_mean = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer_mean, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

preprocessor_knn = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer_knn, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

preprocessor_iterative = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer_iterative, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Modelos completos con diferentes preprocesadores
model_mean = Pipeline(steps=[
    ('preprocessor', preprocessor_mean),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

model_knn = Pipeline(steps=[
    ('preprocessor', preprocessor_knn),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

model_iterative = Pipeline(steps=[
    ('preprocessor', preprocessor_iterative),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Evaluar los modelos con validación cruzada
X = df.drop('target', axis=1)
y = df['target']

print("\nEvaluación de modelos con diferentes estrategias de imputación:")
print("Mediante validación cruzada (5-fold) - R² Score")

scores_mean = cross_val_score(model_mean, X, y, cv=5, scoring='r2')
print(f"Imputación con media: {scores_mean.mean():.4f} ± {scores_mean.std():.4f}")

scores_knn = cross_val_score(model_knn, X, y, cv=5, scoring='r2')
print(f"Imputación con KNN: {scores_knn.mean():.4f} ± {scores_knn.std():.4f}")

scores_iterative = cross_val_score(model_iterative, X, y, cv=5, scoring='r2')
print(f"Imputación iterativa: {scores_iterative.mean():.4f} ± {scores_iterative.std():.4f}")

# Comparar los resultados
plt.figure(figsize=(10, 6))
models = ['Media/Moda', 'KNN', 'Iterativo (RF)']
scores = [scores_mean.mean(), scores_knn.mean(), scores_iterative.mean()]
errors = [scores_mean.std(), scores_knn.std(), scores_iterative.std()]

plt.bar(models, scores, yerr=errors, alpha=0.7, capsize=10, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.ylim([0, 1])
plt.ylabel('R² Score')
plt.title('Comparación de Estrategias de Imputación en Pipeline')
plt.grid(axis='y', alpha=0.3)
plt.show()

#-----------------------------------------------------------------
# 9. CONCLUSIONES Y RECOMENDACIONES FINALES
#-----------------------------------------------------------------

print("""
\n** CONCLUSIONES Y RECOMENDACIONES FINALES **

1. La imputación de datos es un paso crítico en el preprocesamiento que puede tener un impacto significativo
   en el rendimiento de los modelos predictivos.

2. No existe un método de imputación "universal" que sea óptimo para todos los casos:
   - El mejor método depende del patrón de valores faltantes (MCAR, MAR, MNAR)
   - La estructura de los datos y correlaciones entre variables
   - La cantidad de valores faltantes
   - El objetivo del análisis

3. Resultados principales observados:
   - Para datos con pocos valores faltantes, los métodos simples (media, mediana) pueden ser suficientes
   - Los métodos basados en KNN funcionan bien para capturar relaciones entre observaciones similares
   - Los métodos iterativos basados en modelos predictivos (como Random Forest) suelen ofrecer las mejores
     imputaciones pero con mayor costo computacional
   - La imputación multivariante es crucial cuando existen fuertes correlaciones entre variables

4. Recomendaciones prácticas:
   - Siempre comience con un análisis exhaustivo de los valores faltantes
   - Compare múltiples métodos de imputación y evalúe su impacto
   - Utilice pipelines para integrar la imputación en el flujo de trabajo completo
   - Considere crear indicadores de imputación para valores faltantes informativos
   - Para análisis críticos, considere técnicas de imputación múltiple para capturar la incertidumbre

5. Áreas de desarrollo futuro:
   - Imputación con redes neuronales y técnicas de aprendizaje profundo
   - Métodos bayesianos para imputación con estimación de incertidumbre
   - Técnicas específicas para datos de alta dimensionalidad
   - Herramientas automatizadas para selección de métodos de imputación
""")

# Mostrar una tabla comparativa final de los métodos
methods_comparison = pd.DataFrame({
    'Método': ['Eliminación de filas/columnas', 'Media/Mediana/Moda', 'Interpolación',
             'Forward/Backward Fill', 'KNN', 'Regresión/Clasificación', 'MICE', 'PCA'],
    'Complejidad': ['Baja', 'Baja', 'Media', 'Baja', 'Media', 'Alta', 'Alta', 'Alta'],
    'Preserva distribución': ['Alta', 'Baja', 'Media', 'Media', 'Media-Alta', 'Alta', 'Alta', 'Media'],
    'Preserva relaciones': ['Alta', 'Baja', 'Baja-Media', 'Media', 'Alta', 'Alta', 'Alta', 'Alta'],
    'Uso recomendado': ['MCAR, <5% faltantes', 'MCAR, datos simples', 'Series temporales',
                       'Datos secuenciales', 'MAR, datos agrupados', 'MAR/MNAR, relaciones complejas',
                       'MAR/MNAR, múltiples variables', 'Datos de alta dimensión']
})

print("\nResumen comparativo de métodos de imputación:")
print(methods_comparison)

print("\n¡Gracias por utilizar este notebook sobre Métodos de Imputación!")
```
